# 2. Отображения. Обратное отображение

Пусть $ X, Y $ - два множества. Отображением $ f $ множества $ X $ в множество $ Y $ называется закон, посредством которого произвольному элементу $ x \in X $ ставится в соотвенствие однозначно определённый элемент $ y \in Y $. При этом элемент $ y $ называется образом элемента $ x $, а элемент $ x $ - прообразом элемента $ y $. Символически отображение записывается в виде $ f: X \to Y $ или $ X \overset{f}{\to} Y $. Запись $ y = f(x) $ или $ x \mapsto y $ означает, что элемент $ x $ при отображении $ f $ переходит в элемент $ y $.

Также отображение называют преобразованием множества в множество, оператором, действующим из множества в множество, млм функцией, определённой на множестве со значениями другого множества.

Полным прообразом элемента $ y \in Y $ называетсямножество $$ f^{-1}(y) = \{ x \in X | f(x) = y \} $$

Образом отображения $ f $ нызвается множество $$ im \space f = \{ y = f(x) | x \in X \} $$  
Также вместо символа $ im \space f $ используется символ $ f(X) $

Отображение $ f: X \to Y $ называется:
- инъективным, если $ x_1 \neq x_2 \Rightarrow f(x_1) \neq f(x_2) $
- сюрьективным или отображением на, если $ im \space f = Y $ при любом $ y \in Y $ имеет хотя бы одно решение
- биективным или взаимно однозначным, если оно и инъективно и сюрьективно

Тождественным отображением на множестве $ X $ называется отображение $ e_X: X \to X $, которое переводит каждый элемент $ x \in X $ в себя

### Произведение отображений

Произведением отображений $ f: X \to Y $ и $ g: Y \to Z $ называется отображение $ f_g: X \to Z $, определённое правилом $$ f_g(x) = f(g(x)) \space\space \forall x \in X $$

#### Свойства произведений

1. $ fe_x = f, e_yf = f \space\space \forall f: X \to Y $
2. Произведение отображений ассоциативно
3. Произведение ассоциативных (инъективных, биективных) отображений ассоциативно (инъективно, биективно соотв.)

### Обратное отображение

Пусть $ f: X \to Y $. Отображение $ f^{-1}: Y \to X $ называется обратным к $ f $, если $$ f^{-1}f = e_x, \space\space ff^{-1} = e_y $$  
Если выполнено лишь одно из равенств, то такое отображение называется левым или правым обратным отображением.

### Теорема 1 (критерий обратимости)

Отображение обратимо тогда и только тогда, когда оно биективно.

### Лемма

Если $ g: X \to Y, f: Y \to X $ и $ f_g = e_x $, то $ g $ инъективно, а $ f $ - сюрьективно

### Доказательство леммы

Действительно, если $ g $ не инъективно, то существуют элементы $ x_1, x_2 \in X $ такие, что $ x_1 \neq x_2 $, а $ g(x_1) = g(x_2) $. Тогда $ x_1 = e_x(g_1) = fg(x_1) = f(g(x_1)) = f(g(x_2)) = fg(x_2) = x_2 $. Значит $ g $ инъективно. Далее, если $ x $ - произвольный элемент $ X $, то $ x = e_x(x) = fg(x) = f(g(x)) = f(y) $, где $ y = g(x) \in Y $. Это доказывает сюрьективность $ f $

### Доказательство теоремы. Необходимость

Пусть $ f $ обратимо. Тогда из определения обратного отображения и из теоремы 1 слудет, что $ f $ и инъективно, и сюрьективно, т.е. $ f $ биективно.

### Достаточность

Пусть $ f $ биективно. Тогда $ \forall y \in Y $ существует единственный прообраз $ x \in X $. Построим отображение $ g: Y \to X $, положив $ g(y) = x $. Тогда $ \forall y \in Y $ имеем, что $ (fg)(y) = f(g(y)) = f(x) = y $, т.е. $ fg = e_Y $, а $ \forall x \in X $ имеем $ gf(x) = g(f(x)) = g(y) = y $, т.е. $ gf = e_x \Rightarrow g = f_{-1} $

Свойства отображения: 
1. Обратное отображение единственно
2. Произведение обратимых отображений обратимо, при этом $ (fg)^{-1} = g^{-1}f^{-1} $
