# 10. Изоморфизм групп

Две группы $ G_1, G_2 $ с операциями $ *_1, *_2 $ называют изоморфными, если существует биективное отображение $ f: G_1 \to G_2 $, которое сохраняет групповую операцию. Т.е. $ f(a *_1 b) = f(a) *_2 f(b) \space\space\space\space \forall a, b \in G_1 $. Обозначение: $ G_1 \cong G_2 $. Само отображение $ f $пр этом называют изоморфизмом.

Простейшие свойства изоморфизма:
1. Отношение изоморфизма является отношением эквивалентности на множестве всех групп
2. В изоморфных группах $ G1, G_2 $ образ и прообраз единицы являются единицей
3. В изоморфных группах $ G_1, G_2 $ образ и прообраз обратного элемента являются обратным элементом. $ f(a^{-1}) = (f(a))^{-1} $

### Теорема 1 (теорема Кэли)

Любая конечная группа изоморфна некоторой группе перестановок на множестве своих элементов

### Теорема 2

Все бесконечные циклические группы изомофны аддитивной группе целых чисел

### Доказательство

Соответствие $ ak \mapsto k $ будет взаимно одназначным отображением $ G \to \mathbb{Z} $

### Следствие

Все бесконечные циклические группы изоморфны между собой.

Изоморфное отображение группы $ G $ на смебя называется автоморфизмом.
