# 25. Основная теорема алгебры. Разложение многочлена на линейные множители

### Алгебраическая замкнутость поля

Поле $ P $ называется алгебраически замкнутым , если любой многочлен $ f(x) \in P[x] $ степени $ n \geqslant 1 $ обладает в $ P $ хотя бы одним корнем. Очевидно, что поле действительных чисел не является замкнутым, т.к. многочлен $ x^2 + 1 $ не имеет действительных корней.

### Основная теорема алгебры

Поле $ C $ комплексных чисел алгебраически замкнуто.

### Доказательство

#### Лемма 1

Пусть $ f(z) = a_1z + a_2z^2 + \dots + a_n z^n $ - многочлен с нулевым свободным членом. Тогда для любого $ \varepsilon > 0 $ найдётся $ \delta < 0 $ такое, что для всех $ z $, для которых $ |z| < \delta $ выполняется неравенство $ |f(z)| < \varepsilon $

#### Лемма 2

Многочлен $ f(z) = a_0 + a_1z + \dots + a_n z^n $ есть непрерывная функцияя во всех точках комплексной плоскости.

#### Лемма 3

Модуль многочлена есть непрерывная функция.

#### Лемма 4

Если $ f(z) $ - многочлен чтепени $ n \geqslant 1 $, то для любого $ M > 0 \space\space\space\space \exists R > 0 $ такое, что для всех $ z $, для которых $ |z| > R $, выполняется неравенство $ |f(z)| > M $

#### Лемма 5

Из любой ограниченной последовательности можно выделить сходящуюся подпоследовательность.

#### Лемма 6

Точная нижняя грань модуля многочлена достижима, т.е. существует число $ z_0 $ такое, что $ |f(z_0)| \leqslant |f(z)| $ при всех комплексных $ z $

#### Лемма 7 (лемма Даламбера)

Если $ f(z) $ - многочленстепени $ n \geqslant 1 $ и $ f(z_0) \neq 0 $, то найдётся число $ z_1 $ такое, что $ |f(z_1)| \leqslant |f(z_0)| $

### Доказательство осносной теоремы

Пусть $ f(z) $ - произвольный многочлен степени $ n \geqslant 1 $ над полем $ \mathbb{C} $ от комплексной переменной $ z $. Согласно лемме 6 множество всевозможных значений $ |f(z)| $  имеет точную нижнюю грань $ m $, которая достигается в некоторой точке $ z_0 $, так что $ |f(z_0)| = m $. Тогда $ f(z_0) = 0 $, так как в противном случае, если $ |f(z_0)| \neq 0 $, то согласно лемме 7 найдётся точка $ z_1 $, для которой $ |f(z_1)| < |f(z_0)| = \inf|f(z)| $, что невозможно. Таким образом, $ z_0 $ - корень $ f(z) $ и поле $ \mathbb{C} $ комплексных чисел алгебраически замкнуто.

### Теорема 2

Для любого многочлена $ f(z) = \sum_{k = 0}^n a_k z^k \in \mathbb{C}[z] $ степени $ n \geqslant 1 $ существубт числа $ c_1, c_2, \dots, c_n \in \mathbb{C} $ такие, что  

$$ f(z) = a_n(z - c_1)(z - c_2)\cdots(z - c_n) $$

Это разложение эдинственно с точностью до порядка сомножителей.

### Доказательство

Большое и страшное. Моих сил сейчас не хватит на него и расписание последующих билетов
