# 22. Наибольший общий делитель двух многочленов, его свойства. Алгоритм Евклида

Многочлен $ \varphi(x) $ называется общим делителем двух многочленов $ f(x) $ и $ g(x) $, если он является делителем каждого из них. Если у многочленов нег других делителей, кроме многочленов нулевой степени, они называются взаимно простыми.

Многочлен $ d(x) $ называется наибольшим общим делителем (НОД) ненулевых многочленов $ f(x) $ и $ g(x) $, если:
1. $ d(x) $ - общий делитель $ f(x) $ и $ g(x) $
2. $ d(x) $ делится на любой общий множитель $ f(x) $ и $ g(x) $

Обозначение: $ НОД(f, g) $. Очевидно, что если $ d(x) = НОД(f, g) $, то $ \alpha d(x) = НОД(f, g) \space \forall \alpha \neq 0 $

### Теорема 1

Для любой пары $ f(x), g(x) \in P[x] $ существует НОД. Он определён однозначно с точностью до множителя нулевой степени.

### Доказательство

Приведём описание алгоритма построения $ НОД(f, g) $, называемого алгоритмом Евклида или алгоритмом последовательного деления. Он состоит в следующем.

Выполним цепочку делений с остатками согласно теореме о делении многочлена:

$$ \begin{cases} 
f(x)         &= g(x)q_(x) + r_(x) \newline 
g(x)         &= r_1(x)q_1(x) + r_2(x) \newline 
r_1(x)       &= r_2(x)q_3(x) + r_3(x) \newline 
\dots \newline 
r_{k - 1}    &= r_{k - 2}(x) q_{k - 1}(x) + r_{k - 1}(x) \newline 
r_{k - 2}    &= r_{k - 1}(x)q_k(x) + r_k(x) \newline 
r_{k - 1}(x) &= r_k(x)q_{k + 1}(x) \end{cases}, 
\space \space \space \deg g > \deg r_1 > \deg r_2 > \dots > \deg r_k  $$

Степени остатков понижа.тся, поэтому процесс оборвётся в тот момент, когда деление выполнится нацело. Пусть $ r_k(x) $ - последний отличный от нуля остаток. Покажем, что $ r_k(x) = НОД(f, g) $. Действительно, просматривая все равенства снизу вверх, заключаем, что $ r_k(x) $ является делителем $ r_{k - 1}(x), r_{k - 2}(x), \dots, r_1(x), g(x), f(x) $, т.е. общим делителем $ f(x) $ и $ g(x) $. Просматривая равенства сверху вних, заключаем, что все остатки $ r_1(x), r_2(x), \dots, r_k(x) $ делятся на любой общий делитель $ f(x) $ и $ g(x) $.

Докажем вторую часть теоремы. Положим $ d_1(x) = НОД(f, g), d_2(x) = НОД(f, g) $. Тогда согласно определению $ НОД(f, g), d_1(x) = d_2(x)q_1(x), d_2(x) = d_1(x)q_2(x) $ и, с учётом равенства степеней произведения многочленов, $ \deg g_1 \geqslant \deg g_2, \deg g_2 \geqslant \deg g_1 \Rightarrow \deg g_1 = \deg g_2 $ и многочлены $ d_1(x) $ и $ d_2(x) $ отличаются лишь множителями нулевой степени.
