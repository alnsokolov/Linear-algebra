# 13. Поле. Характеристика поля. Алгебраическое расширение поля

Полем называется коммутативное кольцо с единицей, содержащее не менее двух элементом, в котором каждый элемент отличный от нуля имеет обратный элемент.

Простейшие свойства поля:
1. Поле обладает всеми свойствами кольца
2. В нём нет делителей нуля, так как если $ ab = 0, a \neq 0 $, то $ a^{-1}(ab) = b = 0 $
3. В поле $ P $ множество всех ненулевых элементов оразует мультипликативную коммутативную группу, и поэтому в поле:
   - существует, и притом единственная, единица, причём $ 1 \neq 0 $
   - для любого элемента $ a \neq 0 $ существует и притом единственный, обратны элемент
   - $ \forall a, b \in P \space\space \exists! x: ax = b $, при этом $ x = a^{-1}b = ba^{-1} $. Этот элеммент называется частным от деления $ b $ на $ a $ и обозначается как частное у простых чисел
4. В поле сохраняются все обычные правтла обращения с дробями:
   - $ \frac{a}{b} = \frac{c}{d} \Leftrightarrow ad = bc $
   - $ \frac{a}{b} \pm \frac{c}{d} = \frac{ad \pm bc}{bd}; \frac{a}{b} \frac{c}{d} = \frac{ac}{bd} $
   - $ \frac{-a}{b} = \frac{a}{-b} = -\frac{a}{b} $

Таким образом все правила и формулы элементарной алгебры полностью сохраняются в любом поле, т.к. в их основе лежат одни и те же операции сложения, вычитания, умножения и деления. 

Элементы поля называют числами.

#### Расширение поля

Подмножество $ F $ поля $ P $ называется подполем поля $ P $, если оно само является полем относительно операций, оперделённых в $ P $.При этом говорят, что поле $ P $ является расширением поля $ F $, а поле $ F $ вложено в поле $ P $.

Любое поле, элементами которого являются действительнае числа с операциями сложениея, вычитания, умножения и деления действительных чисел, будем называть числовым полем.

### Теорема 1

Поле рациональных чисел вложено в любое числовое поле.

### Доказательство

В поле содержатся 0, 1, а также все натуральные числа, получаемые сложением 1 с самой собой несколько раз подряд. С помощью вычитаний можно получить все целые числа, а с помощью частного - все рациональные.

### Характеристика поля

Существуют поля, в которых некоторое целое, кратное $ 1 $, т.е. $ n1 = \underset{n}{\underbrace{1 + 1 + \dots + 1}} $, равно 0. Наименьшее натуральное число, обладающее этим свойством называется характеристикой поля. Если указанное свойство не имеет места ни для какого натуральногто числа $ n $, то говорят, что такое поле имеет характеристику $ 0 $.  
Примерами полей с характеристикой $ p \geqslant 1 $ служат все конечные поля. В самом деле, конечное поле является конечной аддитивной группой, при этом характеристика поля $ p \geqslant 1 $, так как $ p $ совпадает с порядком элемента $ 1 $, являющимся делителем порядка самой группы.

### Теорема 2

Характеристикой поля может быть либо 0, либо простое число

### Доказательство

Пусть $ n $ - характеристика поля и $ n = mk, \space\space\space\space m, n \in \mathbb{N}, m < n, k < n $ - составное число. Тогда в силу дистрибутивности $$ \underset{n}{\underbrace{1 + 1 + \dots + 1}} = (\underset{m}{\underbrace{1 + 1 + \dots + 1}})(\underset{k}{\underbrace{1 + 1 + \dots + 1}}) = 0 $$ Т.к. в поле нет делителей нуля, то либо $ \underset{m}{\underbrace{1 + 1 + \dots + 1}} = 0 $, либо $ \underset{k}{\underbrace{1 + 1 + \dots + 1}} = 0 $, т.е. $ n $ - не наименьшее натуральное число, обладающее указанными свойствами
