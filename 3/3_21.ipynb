# 21. Кольцо многочленов. Деление многочленов

Пусть $ P $ - поле. Многочленом $n$-й степни от переменной $x$ над полем $ P $ называется выражение $$ a_0 + a_1 x + \dots + a_n x^n, $$ где $ a_i, i = \overline{0, n} $ - фиксированные числа из поля $ P $ и $ a_n \neq 0 $. Эти числа называются коэффицентами многочлена, а число $ a_n $ - страшим коэффицентом. Число $ 0 \in P $ по определению считается многочленом с нулывым коэффицентом и называется нулевым многочленом. Степень нулевого многочлена не определена.

Многочлен обозначается символом $ f(x) $ или $ f_n(x) $, степень многочлена $ f(x) $ - символом $ deg \space f(x) $, множество всех многочленов от переменной $ x $ над полем $ P $ - символом $ P[x] $. Итак $$ f(x) = \sum_{k = 0}^n \limits a_kx^k \in P[x], \space deg \space f = n $$ Два многочлена называются равными (обозначение $ f(x) = g(x) $), если $$ deg \space f = deg \space g = n, \space\space a_i = b_i, i = \overline{0, n} $$

Введём на множестве $ P[x] $ алгебраические операции.  

Суммой многочленов $ f(x) = \sum_{k = 0}^n a_k x^k, g(x) = \sum_{k = 0}^s b_k x^k $ будем называть многочлен

$$ h(x) = f(x) + g(x) = \sum_{k = 0}^{(max(n, s))} c_k x^k, \space\space c_k = a_k + b_k $$ Здесь недостающие коэффиценты $ a_k $ и $ b_k $ заменяются нулями.

Из определения вытекают следующие факты:
1. $\forall f(x) \in P[x]$ $$ f(x) + 0 = 0 + f(x) = f(x) $$
2. $ \forall f(x) \neq 0, g(x) \neq 0, f(x) + g(x) \neq 0 $ $$ deg (f + g) \leqslant max(deg \space f, deg \space g) $$
3. Если $ f(x), g(x) \in P[x] $, то $ f(x) + g(x) \in P[x] $, т.е. сложение многочленов является алгебраической операцией на множестве $ P[x] $

Произведением многочленов $ f(x) = \sum_{k = 0}^n a_k x^k, g(x) = \sum_{k = 0}^s b_k x^k $ будем называть многочлен

$$ h(x) = f(x)g(x) = \sum_{k = 0}^{n + s} c_k x^k, \space\space c_k = \sum_{i + j = k} a_i, b_j, \space k = \overline{0, n + s} $$

Из определения вытекают следующие факты:
1. Произведение ненулевых многочленов не может быть нулевымЮ при этом $$ deg \space fg = deg \space f + deg \space g  $$
2. Если $ f(x), g(x) \in P[x] $, то $ f(x)g(x) \in P[x] $, то есть умножение является алгебраической операцией на $ P[x] $
3. Операция умножения многочленов порождает операцию умножения многочлена на элемент из поля $ P $ как частный случай умножения многочленов:

$$ \alpha f(x) = \sum_{k = 0}^n \alpha a_k x^k $$

Таким образом на множестве $ P[x] $ определён и внешний закон композиции.

### Теорема 1

Множество $ P[x] $ всех многочленов над полем $ P $ является кольцом с единицей и без делителей нуля.

### Доказательство

Доказывается рассмотрением аксиом для кольца.

## Деление многочленов

Кольцо $ P[x] $ всех многочленов над полем $ P $ по своим свойствам близко к кольцу $ \mathbb{Z} $ всех целх чисел. Эта аналогия проявляется и в том, что для многочленов, как и для целых чисел существует понятие деления нацело, деления с состатком, делителя, наибольшего общего делителя и т.п.

### Теорема 2

Для любых двух многочленов $ f(x), g(x) \in P[x] $, где $ g(x) \neq 0 $, существует, и ритом единственная, пара многочленов $q(x), r(x) \in P[x] $, такая что^

$$ f(x) = g(x)q(x) + r(x), $$ где либо $ r(x) = 0 $, либо $ \deg r < \deg g $

### Доказательство

Докажем первую часть теоремы. Будем считать, что $ f(x) \neq 0 $. Пусть $ f(x) = \sum_{k = 0}^n a_k x^k, g(x) = \sum_{k = 0}^s b_k x^k, \deg r < \deg g $. Без ограничения общности скажем, что $ n \geqslant s $. Применим метод математической индукции по степени $ n $ многочлена $ f(x) $, считая $ g(x) $ фиксированным.

1. Пусть $ n = 0 $. Тогда $ s = 0 $ и $ q(x) = a_0b_0^{-1}, r(x) = 0 $
2. Пусть теперь теорема верна для любого многочлена степени меньшей $ n $. Докажем её для многочлена степени $ n \geqslant s $. Воспроизведём первый шаг алгоритма деления многочленов с действительными коэффицентами, т.е. построим одночлен $ \frac{a_n}{b_s}x^{n - s} $ и составим разность

$$ f_1(x) = f(x) - \frac{a_n}{b_s}x^{n - s} g(x) $$

Либо многочлен $ f_1(x) $ равен нулю, либо $ \deg f_1 < n $. В первом случае можно положить $ q(x) = \frac{a_n}{b_s}x^{n - s}, r(x) = 0 $. Во втором случае для многочлена $ f_1(x) $ по индуктивному предположению найдутся многочлены $ q_1(x), r(x) $ такие, что $ f_1(x) = g(x)q_1(x) + r(x) $, гдк либо $ r(x) = 0 $, либо $ \deg r < \deg g $. Тогда $ f(x) = \frac{a_n}{b_s}x^{n - s}g(x) + g(x)q_1(x) + r(x) $. Положив $ q(x) = \frac{a_n}{b_s}x^{n - s} + q_1(x) $, приходим к паре многочленов $ q(x), r(x) \in P[x] $, удовлетворяющей условиям теоремы

Остаётся доказать единственность. Пусть существует ещё одна пара многочленов $ q_1(x), r_1(x) \in P[x] $, удовлетворяющая условиям теоремы. Тогда $ f(x) = g(x)q(x) + r(x) $ и $ f(x) = g(x)q_1(x) + r_1(x) $, т.е

$$ g(x)(q(x) - q_1(x)) = r_1(x) - r(x) $$

Если $ r_1(x) - r(x) \neq 0 $, то и $ q(x) - q_1(x) \neq 0 $, при этом степень правой части равенства меньше $ s $ , а левой части - не меньше $ s $, что невозможно. Следовательно $ r_1(x) - r(x) = 0, q(x) - q_1(x) = 0 \Leftrightarrow r_1(x) = r(x), q_1(x) = q(x) $

Многочлен $ q(x) $ называется частным (или неполным частным) от деления $ f(x) $ на $ g(x) $, а $ r(x) $ - остатком от этого деления. При этом, если $ r(x) = 0 $, то говорят, что $ f(x) $ делится на $ g(x) $, при этом $ g(x) $ называется делителем $ f(x) $.
