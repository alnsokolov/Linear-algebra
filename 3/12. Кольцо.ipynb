# 12. Кольцо

Допустим у нас есть множество $ K $, наделённое двумя операциями. Одну из них будем называть сложением и обозначать символом $ + $, другую - умножением и обозначать символом $ \cdot $. Такое множество называется кольцом, если $ \forall a, b \in K $ выполняются аксиомы:
1. $ a + b = b + a $
2. $ (a + b) + c = a + (b + c) $
3. $ \exists 0 \in K:  a + 0 = 0 + a = a $
4. $ \forall a \in \space\space K \exists -a \in K: a + (-a) = (-a) + a = 0 $
5. $ (ab)c = a(bc) $
6. $ (a + b)c = ac + bc, a(b + c) = ab + ac $

Кольцо называется коммутативным, если операция умножения коммутативна. Колько называется кольцом с единицей, если упножение обладает нейтральным элементом. Из аксиом кольца следует, что оно является аддитивной абелевой группой.

Простейшие свойства кольца:

1. Кольцо обладает всеми свойства аддитивной абелевой группы:
    - существуе, и притом единственный, нулевой элемент $ 0 $
    - для каждого элемента $ a $ существует, и притом единственный, обратный элемент $ -a $
    - для любых элементов $ a, b \in K $ существуе, и притом единственное, решение уравнения $ x + a = b $; это решение называется разностью элементов $ b $ и $ a $ и обозначается $ b - a $. Итак $ b - a = b + (-a) $
    - Определены целые кратные элемента: 
    $$ na = \begin{cases} \underset{n}{\underbrace{a + a + \dots + a}}, & n > 0 \newline 0, & n = 0 \newline m(-a) = -(ma) & m = -n, n < 0 \end{cases} $$
2. В кольце умножение дистрибутивно относительно вычитания
3. В кольце для любого элемента $ a0 = 0a = 0 $
4. В кольце для любых элементов $ (-a)b = a(-b) = -ab \Rightarrow (-a)(-b) = ab $
5. В кольце с единицей для любого элемента $ a: \space\space (-1)a = a(-1) = -a $
6. В кольце с единицей, содержащем не менее двух элементов $ 1 \neq 0 $
7. В кольце с единицей множество обратимых элементов образует мультипликативную группу
