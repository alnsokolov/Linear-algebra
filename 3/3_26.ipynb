# 26. Каноническое разложение над полем комплексных чисел. Кратность корня

### Теорема 1

Для любого многочлена $ f(z) = \sum_{k = 0}^n a_k z^k \in \mathbb{C}[z] $ степени $ n \geqslant 1 $ существубт числа $ c_1, c_2, \dots, c_m \in \mathbb{C} $, где $ c_i \neq c_j $ при $ i \neq j $ и числа $ k_1, k_2, \dots, k_m \in \mathbb{N} $, где $ k_1 + k_2 + \dots + k_m = n $, такие, что

$$ f(z) = a_n(z - c_1)^{k_1} \dots (z - c_m)^{k_m} $$

Это разложение единственно с точностью до порядка сомножителей

### Доказательство

Рассмотрим разложение $ a_n(z - c_1)(z - c_2)\cdots(z - c_n) $ для многочлена $ f(z) $. Среди чисел $ c_1, \dots, c_n $ могут быть одинаковые. Будем считать, что среди чисел $ c_1, \dots, c_m $ в различны, а каждое из чисел $ a_{m + 1}, \dots, a_n $ равно одному из вышеупомянутых. Объединим одинаковые сомножители в этом проихведении, получим  

$$ f(z) = a_n(z - c_1)^{k_1} \dots (z - c_m)^{k_m} $$  

где $ c_i \neq c_j $ при $ i \neq j $ по построиению, а $ k_1 + k_2 + \dots + k_m = n $ как число всех сомножителей. Есдинственность разложения следует из теоремы о разложении многочлена.

Это разложение называеся каноническим разложением многочлена над полем комплексных чисел.

Число $ k_i $ в каноническом разложении многочлена называется кратностью корня $ c_i $. Если $ k_i = 1 $, корень $ c_i $ называется простым, если $ k_i > 1 $ - кратным.
