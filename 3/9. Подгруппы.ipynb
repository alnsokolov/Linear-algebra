# 9. Подгруппы. Смежные классы. Нормальные делители

Пусть $ H $ - подгруппа группы $ G $, $ a $ - элемент $ G $. Множество $ aH $ называется левым смежным классом группы $ G $ по подгруппе $ H $, порождённым элементом $ a $, а $ Ha $ - левым смежным классом. Из поределения вытекают следующие простейшие свойства смежных классов:

1. $ a \in aH, a \in Ha $, так как подгруппа $ H $ содержит нейтральный элемент
2. Смежный класс состоит из элементов группы, причёт каждый элемент шруппы входит в какой-нибудь смежный класс
3. Подгруппа $ H $ является одним из смежных классов ($ H = He = eH $)
4. В абелевой группе $ aH = Ha, \space\space\space\space \forall a \in G $

### Теорема 1

Смежный класс порождается любым своим элементом

### Доказательство

Надо показать, что если $ g \in aH $, то $ aH = gH $. Пусть $ g = ah_1, h_1 \in H $. Тогда $ \forall h \in H $ имеем $ ah = (ah_1)(h_1^{-1}h) = gh_2 $, где $ h_2 = h_1^{-1}h \in H \Rightarrow aH \subset gH $, и с учётом вложения в другую сторону получено требуемое равенство. Доказательство может быть использовано и для правого класса

### Теорема 2

Любые два левых(правых) смежных класса либо совпадают, либо не пересекаются.

### Доказательство

Если два смежных класса имеют хотя бы один смежный элемент, то они им порождаются, а, следовательно, и совпадают.

## Нормальные дилители

Подгруппа $ H $ называется нормальным делителем, если для любого элемента $ a \in G $ $$ aH = Ha $$ т.е. если правый и левый смежные классы совпадают.

Очевидно, что в абелевой группе любая подгруппа является нормальным делителем. Тривиальные группы в каждой группе также являются нормальными делителями.

Элементы $ a, b \in G $ называется сопряжёнными, если $ \exists c \in G: a = c^{-1}bc $

### Теорема 1

Подгруппа $ H $ группы $ G $ является нормальным делителем тогда и только тогда, когда она вместе с каждым элементом содержит все сопряжённые с ним элементы

### Доказательство

$ \dots $

### Теорема 2

Смежные классы по нормальному делителю образуют группу относительно умножения подмножества группы.

### Доказательство

$ \dot $

Группа смежных классов группы $ G $ по нормальному делителю $ H $ называется фактор-группой группы $ G $ по подгруппе $ H $.
