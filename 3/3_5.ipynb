# 5. Подгруппы. Симметрическая и знакопеременная группы

Непустое подмножество $ H $ группы $ G $ является подгруппой группы $ G $, если оно само является группой относительно алгебраической операции в $ G $.

Простейшими ппримерами подгрупп являются нейтральный элемент и сама группа. Они называются тривиальными.

### Теорема 1

Подмножество $ H $ группы $ G $ является подгруппой этой группы тогда и только тогда, когда выполняются следующие импликации:

$$ \begin{align} &1. a, b \in H \Rightarrow ab \in H \newline &2. a \in H \Rightarrow a^{-1} \in H \end{align} $$

### Доказательство

Необходимость очевидна, она вытекает из того, что $ H $ - группа.

Достаточность. Из первого условия следует, что алгебраическая операция в $ G $ является алгебраической операцией в $ H $. Что же касается аксиом, то необходимо проверить только аксиому единицы. Пусть $ a \in H $; тогда, согласно условиям 1 и 2, $ a^{-1} \in H $ и $ aa^{-1} = e \in H $

### Коннечная группа

Группа с конечным числом элементов называется конечной группой. Число элементов такой группы называется её порядком и обозначается символом $ card \space G $. Бесконечную группу называют группой бесконечного порядка.

### Симметрическая группа $n$-го порядка

Множество $ S_n $ всех подстановок $ n $-го порядка образует мультипликативную группу как множество всех биективных отображений множества на себя. Эта группа называется симметрической группой $n$-го порядка, её порядок равен $ n! $

### Знакопеременная группа $n$-го порядка

Множество $ A_n $ всех чётных подстановок образует мультипликативную группу. Для того, чтобы в этом убедиться, достаточно показать, что оно является подмножеством $ S_n $. Группа $ A_n $ называется знакопеременной группой $ n $-го порядка. Её порядок равен $ \frac{n!}{2}, \space n \geqslant 2 $
