# 6. Группа невырожденных матриц. Группа невырожденных треугольных матриц. Группа ортогональных матриц

### Группа невырожденных матриц

Множество $ M $ невырожденных матриц $n$-го порядка образует группу относительно ассоциативной операции умножения матриц, т.к. в соответствии с теоремой 37.1, это множество является группой: в нём есть нейтральный элемент (единичная матрица) и обратный элемент (обратная матрица, т.к. матрица невырожденная)

### Группа невырожденных треугольных матриц

Множество $ M $ невырожденных треугольных матриц $n$-го порядка образует мультипликативную группу. В этом можно убедиться, показав, что $ M $ - подгруппа группы невырожденных матриц $n$-го порядка, т.е. проверив все требования теоремы о том, какое подмножество является подгруппой.

1. Если $ A, B $ - верхние треугольные матрица, то $ k $-1 столбец  матрицы $ AB $ является линейной комбинацией первых $ K $ столбцов матрицы $ A $, поэтому все его элементы располооженные ниже $k$-й строки, равны нулю, и, следовательно, $ AB $ - верхняя треугольная матрица
2. Если $ A $ - верхняя невырожденная треугольная матрица, то $k$-й столбец матрицы $ A^{-1} $ является решением системы уравнений $$ Ax = e_k $$ с треугольной матрицей $ A $, у которой все диагональные элементы отличны от нуля; решая эту систему снизу вверх, получаем $$ x_n = 0, x_{n - 1} = 0, \dots, x_{k + 1} = 0 $$ т.е. $ A^{-1} $ - верхняя треугольная матрица4 невырожденность $ A^{-1} $ очевидна.

### Группа ортогональных матриц

Множество $ M $ всех ортогональных матриц $ Q \in \mathbb{R}^{n \times n} $ образует мультипликативную группу, так как
1. произведение ортогональных матриц - ортогональная матрица $$ (Q_1Q_2)^TQ_1Q_2 = Q_2^TQ_1^TQ_1Q_2 = I, \space\space\space\space Q_1Q_2(Q_1Q_2)^T = Q_1Q_2Q_2^TQ_1^T = I $$
2. Обратная к ортогональной матрице - ортогональная матрица

И, согласно теореме о подгруппе, $ M $ - подгруппа группы невырожденных матриц порядка $ n $.
