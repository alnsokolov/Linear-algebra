# 20. Группа корней из единицы. Первообразные корни

## Группа корней $n$-й степени из единицы

Рассмотрим множество $ H_n = \{ \varepsilon_0, \varepsilon_1, \dots, \varepsilon_{n - 1} \} $ всех корней $ n $-й степени из единицы.
1. $ H_n $ - мультипликативная абелева группаЮ как подгруппа мультипликативной группы ненулевых комплексных чисел
2. $ H_n $ - циклическая группа, порождённая корнем $ \varepsilon_1 $, т.к. $ \varepsilon_k = \varepsilon_1^k $

Более того, этой группой исчерпываются по существу все конечные циклические группы порядка $ n $.

### Теорема 1

Все циклические группы порядка $ n $ изоморфмы мультипликативной группе корней $n$-й степени из единицы.

### Доказательство

Пусть $ a $ - образующий элемент циклической группы $ G $. Построим изоморфизм. Каждому элементу $ a^k \in G, 0 \leqslant k < n $ поставим в соответствие корень $ \varepsilon_k n$-й степени из единицы. Это будет взаимно однозначное отображение группы $ G $ на мультипликативную группу $ H_n $, изоморфность которого следует из (39.3)

Корень $ \varepsilon_k $ называеься первообразным корнем $n$-й степени из единицы, если он не является корнем из единицы никакой меньшей степени, чем $ n $. Очевидно, что для любого $ n \geqslant 2 $ число $ \varepsilon_1 $ - первообразный корень. Помимо $ \varepsilon_1 $ могут существовать и другие первообразные корни.

### Теорема 2

Корень $ \varepsilon n$-й степени из единицы является первообразным корнем тогда и только тогда, когда его степени $ \varepsilon^k, \space k = \overline{0, n - 1} $ различны.

### Доказательство. Необходимость

Пусть $ \varepsilon^k = \varepsilon^l $ при $ 0 \leqslant k < l < n - 1 $. Тогда $ \varepsilon^{l - k} = 1 $, где $ 0 \leqslant l - k \leqslant n - 1 $. Следовательно, $ \varepsilon $ является корнем единицы менбшей степени чем $ n $.

### Достаточность

Пусть $ \varepsilon $ - не первообразный корень. Тогда существует $ k < n $ такое, что $ \varepsilon^k = 1 = \varepsilon^0 $ и степени $ \varepsilon^0, \varepsilon^1, \dots, \varepsilon^{n - 1} $ будуь различными.

### Следствия
1. Корень $n$-й степени из единицы является первообразным корнем тогда и только тогда, когда он является образующим циклической группы $ H $ всех корней $n$-й степени из единицы
2. Корень $ \varepsilon_k \in H \space n$-й степени из единицы является первообразным корнем тогда и только тогда, когад $ k $ и $ n $ взаимно просты
