# 4. Группы. Основные свойства

Непустое множество $ G $ с определённой на нём алгебраической операцией $ * $ называется группой, если:

1. Операция ассоциативна $ a * (b * c) = (a * b) * c $
2. Операция обладает нейтральным элементом $ e \in G: a * e = e * a = a \forall a \in G $
3. Для любого элемента $ a \in G $ существует симметричный элемент $ a' \in G: a * a' = a' * a = e $

Обозначение: $ G $ или $ \langle G, * \rangle $

Группа называется абелевой, если определённая на ней опрация является коммутативной.

#### Факты

1. В любой группе существует, и при том единственный, нейтральный элемент
2. В любой группе для каждого элемента существует, и при том единственный, симметричный элемент

Пусть $ G $ - множество, наделённое алгебраической операцией. Элемент $ e_n \in G $ такой, что $ ae_n = a \space\space \forall a \in G $, назовём правой единицей множества $ G $, а $ a_n^{-1} \in G $ такой, что $ aa_n^{-1} = e_n $ - правымобратным к элементу $ a $

### Теорема 1

Множество $ G $ с ассоциативной алгебраической операцией является группой, если оно обладает правой единицей и если каждый элемент $ a \in G $ обладает правым обратным элементом.

### Доказательство

Пусть $ a $ - произвольный элемент множества $ G $ и $ a_n^{-1} $ - правый обратный к нему элемент. Тогда $ aa_n^{-1} = e_n = e_ne_n = e_n(aa_n^{-1}) = (e_na)a_m^{-1} $. Домножим обечасти уравнения на $ a_n $ справа. Получается, что $ a = e_na $, т.е. $ e_n $ является просто единицей множества $ G $. Также $ ea_n^{-1} = a_n^{-1}e = a_n^{-1}(aa_n^{-1}) = (a_n^{-1}a)a_n^{-1} $, т.е. $ e = a_n^{-1}a $

### Теорема 2

Множество $ G $ с ассоциативной алгебраической операцией является группой тогда и только тогда, когда эта операция обладает обратной

### Доказательство. Необходимость

Пусть $ G $ - группа. Покажем, что уравнения $$ \begin{align} ax = b \newline ya = b \end{align} $$ имеют одно решение для любых $ a, b \in G $. Действительно, элемент $ x = a^{-1}b $ является решением первого уравнения, так как $ ax = a(a^{-1}b) = (aa^{-1})b = eb = b $. Это решение едиственно, поскольку если $ x' $ - любой другое решение, то $ ax' = ax $ и $ a^{-1}(ax') = a^{-1}(ax) \Rightarrow x' = x $. Аналогично доказывается единственность решения для второго уравнеиня

### Достаточность

Пусть эти же уравнения имеют единственное решение при любых $ a, b \in G $. Возьмём произвольный элемент $ a \in G $. Уравнение $ ax = a $ имеет единственное решение. Очевидно, оно играет роль правой единицы для элемента $ a $. Обозначим его как $ e_n^a $/. Для любого элемента $ b \in G $ существует элемент $ y \in G $, такой, что $ b = ya $. Тогда $ b = ya = y(ae_n^a) = (ya)e_n^a = be_n^a $. Следовательно $ be_n^a = b \forall b \in G $ и $ e_n^a=e_n $ являетсяправой единицей множества $ G $. Далее уравнение $ ax = e_n $ имеет единственное решение, которое, очевидно, является обратным к элементу $ a $ по отношению к $ e_n $. Отсюда в силу теоремы 1 $ G $ - группа
