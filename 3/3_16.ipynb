# 16. Поле комплексных чисел. Комплексная плоскость

## Понятие комплексного числа

Комплексными числами называются упорядоченные пары $ (a, b) $ вещественных чисел, для которых понятия равенства, суммы, произведения и отождествоения с вещественными числами вводятся слежующим образом:

1. $ (a, b) = (c, d) \Leftrightarrow a = c, b = d $
2. $ (a, b) + (c, d) = (a + c, b + d) $
3. $ (a, b) \cdot (c, d) = (ac - bd, ad + bc) $
4. пара $ (a, 0) $ отождествляется с числом $ a $

Обозначения: $ z = (z, b), \space\space \mathbb{C} $ - множество всех комплексных чисел

### Теорема 1

Множество всех копмлексных чисел является полем.

### Доказательство

Действительно, сложение и умножение являются алгебраическими операциями на множестве $ C $, при том они подчиняютсся всем аксиомам поля. Отметим лишь, что $ 0 = (0, 0); \space 1 = (1, 0); \space -z = (-a, -b); \space z^{-1} = (\frac{a}{a^2 + b^2}, \frac{-b}{a^2 + b^2}) $, если $ z = (a, b) \neq 0 $

### Следствия

1. Для любой пары комплексных чисел существует, и притом единственная, разность
2. Для любой пары комплексных чисел (если второе не равно $ 0 $) существует, и притом единственное, частное

## Алгебраическая форма комплексного числа

Введём обозначение $ i = (0, 1) $. Из аксиом вытекают свойства:
1. $ i^2 = -1 $
2. $ \forall z = (a, b) \in \mathbb{C} \space\space z = a + bi $

Эта форма записи комплексного числа называется алгебраической формой комплексного числа, при этом число $ a $ называется действительной частью числа (обозначается $ Re \space z $), а $ b $ - мнимой ($ Im \space z $ соответственно).

Комплексное число $ \overline{z} = a - bi $ называется сопряжённым к числу $ z = a + bi $

### Теорема 2

Эта операция обладает следующими свойствами:

1. $ \overline{\overline{z}} = z $
2. $ z = \overline{z} \Leftrightarrow z \in \mathbb{R} $
3. $ z + \overline{z} = 2a, \space\space \forall z = a + bi $
4. $ \overline{z}z = a^2 + b^2, \space\space \forall z = a + bi $
5. Операция дистрибутивна относительно всех операций, определённых над комплексными числами

## Комплексная плоскость

Пусть на плоскости выбрана прямоугольная декартова система координат. Поставим в соответствие каждому комплексному числу $ z = a + bi $ точку этой плоскости $ M(a, b) $. Очевмдно, что это соответствие взаимно однозначно.

Плоскость, на которой изображаются комплексные числа называется комплексной плоскостью, её ось абсцисс - вещественной осью, ось ординат - мнимой осью.

При сложении (вычитании) комплексных чисел, складываются (вычитаются) координаты радиус-веторов точек, изображающих сами числа. Поэтому сложение (вычитание) комплексных чисел расносильно сложению (вычитанию) радиус-векторов и может быть выполнено по правилу параллелограмма.
