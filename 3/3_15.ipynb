# 15. Линейное пространство над полем. Число элементов в конечном поле

## Изоморфизм полей и колец

Среди множества колец (и полей) встречаются кольца (и поля), неразличимые с точки зрения свойств, действующих в них алгебраических операций. Эта одинаковость формализуется понятием изоморфизма.  
Два кольца $ K $ и $ K' $ (два поля $ P $ и $ P' $) называются изоморфными, если существует биективное отображение $ \varphi: K \to K' $ (соответственно $ \varphi: P \to P' $), сохраняющие операции:  

$$ \varphi(a + b) = \varphi(a) + \varphi(b), \varphi(ab) = \varphi(a)\varphi(b), \space\space \forall a, b \in K (P) $$  

Нетрудно показать, что
1. Отношение изоморфизма является отношением эквивалентности на множестве всех колец
2. В изоморфных кольцах (и полях):
   - образ и прообраз нуля есть нуль
   - образ и прообраз противоположного элемента есть противоположный элемент
   - образ и прообраз разности есть разность соответствующих элементов
   - образ и прообраз единица (в случае кольца, если оно обладает единицей) есть единица
   - образ и прообраз обратного элемента (в случае кольца, если элемент обладает обратным) есть обратный элемент
3. Свойство кольца иметь или не иметь делители нуля сохраняется в изоморфном кольце

Таким образом, изоморфные кольца и поля могут отличаться друг от друга лишь природой своих элементов, но они тождествены по своим алгебраическим свойствам.

### Теорема 1

Любое поле характеристики $ p \geqslant 1 $ с точностью до изоморфизма является расширением поля вычетов $ \mathbb{Z}_p $.

### Доказательство

В поле характеристики $ p $ элементы, кратные $ 1 $, т.е. $ n1, n = \overline{0, p - 1} $, различны.  Легко показать, что эти элементы образуют подполе. Оно изоморфно полю вычетов $ \mathbb{Z}_p = \{ C_0, \dots, C_{p - 1} \} $: изоморфизм устанавливается отображением $ \varphi(n1) = C_n, n = \overline{0, p - 1} $

### Следствие

Любое конечное поле с точностью до изоморфизма является расширением некоторого поля вычетов.

### Теорема 2

В конечном поле число элементов $ n $ имеет вид $$ n = p^m $$ где $ p $ - простое число, $ m $ - натуральное число.

### Доказательство

Любое конечное поле является полем характеристики $ p \geqslant 1 $. Так как у изоморфных полей одинаковое число элементов, то согласно теореме 1 достатосно показать, что в конечном расширении поля $ \mathbb{Z}_p $ содержится $ n = p^m $ элементов, где $ m $ - натуральное число ($ p $ - простое, так как $ \mathbb{Z}_p $ - поле). Покажем это.

Пусть $ F = \{ a_0, a_1, \dots, a_{k - 1} \} $ - расширение поля $ \mathbb{Z}_p = \{ C_0, \dots, C_{p - 1} \} $. Очевидно, что поле $ F $ является линейным пространством над полем $ \mathbb{Z}_p $. Пусть $ e_0, \dots, e_m $ - базис пространства $ F $. Тогда $ F $ - множество всевозможных линейных комбинаций $ \alpha_1e_2 + \dots + \alpha_m e_m, \space\space \alpha_i \in \mathbb{Z}_p, \space\space i = \overline{1, m} $. Так как для каждого $ \alpha_i $ возможно $ p $ различных значений, то всего элементов будет $ P^m $
