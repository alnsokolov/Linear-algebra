# 3. Алгебраические операции. Обобщённый закон ассоциативности

### Внутренний закон композиции

Внутренним законом композиции (алгебраической операцией) на множестве $ X $ называется отображение $$ *: X \times X \to X $$ т.е. закон, посредством которого любой упорядоченной паре $ a, b \in X $ ставится в соответствие однозначно определённый элемент $ c \in X $. Тот факт, что $ (a, b) \mapsto c $ записывается символически в виде $ a * b = c $. В конкретных случаях вместо символа $ * $ используют символы $ +, -, \times, : $ и др.

Алгебраическая операция $ * $ на множесве $ X $ называется:
- коммутатичной, если $ a * b = b * a \space\space \forall a, b \in X $
- ассоциативной, если $ (a * b) * c = a * (b * c) $

### Обобщённая ассоциативность

Свойство ассоциативности алгебраической операции означает, что результат применения операцрр к трём элементам не зависит от распределения скобок, указывающих на порядок последовательного применения операции к парам элементов. Для ассоциативной операции это же верно и для любого конечного числа $ n $.

### Теорема 1

Если алгебраическая операция $ * $ на множестве $ X $ ассоциативна, то результат её применения не зависит от расстановки скобок, лишь бы скобки объединяли пары элементов, указывая на порядок последовательного применения операции к двум элементам

### Доказательство

Докажем индукцией до $ n $. Для 3 это верно по условию. Пусть утверждение верно для $ k = n $. Пусть в выражении $ a_1, a_2, \dots, a_n $ произвольным образом расставлены скобки. Заметим, что последним шагом будет всегда выполнение операции над двумя элементами $ a_1 * \cdots * a_k $ и $ a_{k + 1} * \cdots a_n, 1 \leqslant k \leqslant n - 1 $. Так как $ k < n $, то оба вырадения не зависят от распределения скобок. Осталось показать, что $$ (a_1 * \cdots * a_k) * (a_{k + 1} * \cdots * a_n) = (a_1 * \cdots * a_m) * (a_{m + 1} * \cdots * a_n) $$, где $ 1 \leqslant m \leqslant n - 1 $. Пусть $ k < m $. Обозначим $ a_1 * \cdots * a_k = a, a_{k + 1} * \cdots * a_n = b, a_{m + 1} * \cdots * a_n = c $. Надо показать, что $ a * (b * c) = (a * b) * c $ верно, что следует из условия.

Элемент $ e \in X $ называется нейтральным элементом множества $ X $ относительно операции $ * $, если $ \forall a \in X $: $$ a * e = e * a = a $$

### Теорема 2

Нейтральный элемент единственнен.

### Доказательство

Допустим это не так и есть $ e_1, e_2 $. Тогда $ e_1 * e_2 = e_1, e_1 * e_2 = e_2 \Rightarrow e_1 = e_2 $
