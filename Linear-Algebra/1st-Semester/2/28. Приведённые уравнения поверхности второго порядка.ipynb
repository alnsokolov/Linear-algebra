# 28. Приведённые уравнения поверхности второго порядка. Метод вращений

### Приведённые уравнения

С помощью ортогонального произведения координат и параллельного переноса уравнение поверхности сторого порядка в пространстве приводится к одному и только одному из пяти простейших уравнений в зависимости от значения $ r = rg \space A $

1. При $ r = 3 $:
   $$ \lambda_1 x^2 + \lambda_2 y^2 + \lambda_3 z^2 + a_0 = 0, \space\space \lambda_1\lambda_2\lambda_3 \neq 0 $$
2. При $ r = 2 $:
   $$ \lambda_1 x^2 + \lambda_2 y^2 + 2b_0z = 0, \space\space \lambda_1\lambda_2b_0 \neq 0 $$  
   
или
$$ \lambda_1 x^2 + \lambda_2 y^2 + c_0 = 0, \space\space \lambda_1\lambda_2 \neq 0 $$
3. При $ r = 1 $:
   $$ \lambda_1 x^2 + 2p_0y + 0, \space\space \lambda_1p_0 \neq 0 $$
   
или
$$ \lambda_1 x^2 + q_0 = 0, \space\space \lambda_1 \neq 0 $$

### Метод вращений

Зададим функцию $ f = a_{11}x^2 + a_{22}y^2 + a_{33}z^2 + 2a_{12}xy + 2a_{13}xz + 2a_{23}yz $

Рассмотрим сферу $ S $, заданную уравнением $ x^2 + y^2 + z^2 = 1 $. Функция $ f $ замкнута и непрерывна на ограниченном множестве $ S $ и, следовательно, достигает на этом множестве максимального значения. Пусть точка макимума функции на сфере $ S $ - $ P $.  

Выполним поворот сситемы координат $ Oxyz $ вокруг $ O $ так. чтобы ось $ Oz' $ совпала с прямой $ OP $, а положительное направление оси - с направлением вектора $ \overrightarrow{OP} $. Получим новую координатную систему $ Ox'y'z' $, в которой функция $ f $ имеет вид $ f = a_{11}'x'^2 + a_{22}'y'^2 + a_{33}'z'^2 + 2a_{12}'x'y' + 2a_{13}'x'z' + 2a_{23}'y'z' $, сфера определяется уравнением $ x'^2 + y'^2 + z'^2 = 1 $, а $ P $ имеет координаты $ (0, 0, 1) $. 

Согласно выбору сситемы координат, максимальное значение функции $ f $ достигается при $ x' = 0, y' = 0, z' = 1 $. Еслли в $ f $ положить $ y' = 0, z' = 1 $, то получится функция $ f_1 $ от одной переменной $ x' $:

$ f_1 = a_{11}'x'^2 + 2a_{13}'x' + a_33' $, которая достигает значения в точке $ P $, т.е. при $ x' = 0 $. Значит $ \frac{df_1}{dx}|_{x' = 0} = 0 \Rightarrow a_{13} = 0 $. Аналогично, если в $ f_1 $ положить $ x' = 0, z' = 1 $, то получим, что $ a_{23}' = 0 $. Таким образом, в системе координат $ Ox'y'z' $ функция $ f_1 $ имеет вид 

$$ f_1 = a_{11}'x'^2 + 2a_{12}'x'y' + a_{22}'y'^2 + a_{33}'z'^2 $$

При повороте сситемы координат $ Ox'y'z' $ вокруг оси $ Oz' $ на угол $ \varphi $ координаты $ x', y', z' $ преобразуются по формулам

$$ \begin{cases} x' = x''cos\varphi - y''sin\varphi \newline y' = x''sin\varphi + y''cos \varphi \newline z' = z'' \end{cases} $$

пр этом функция $ f' $ всё ещё будет иметь вид $ f_1 $, а выбором угла $ \varphi $, для которого $ ctg 2\varphi = (a_{11}' - a_{22}') / (2a_{12}) $ можно добиться того, чтобы коэффицент $ a_{12}' $ стал равен 0.

Таким образом, с помощьюю двух поворотов сситемы координат $ Oxyz $ можно найти систему координат $ Ox''y''z'' $, в которой группа старших членов не содержит попарных произведений $ x'', y'' $ и $ z'' $. Этот метод называется методом вращений квадратичных форм от переменных к главным осям.
