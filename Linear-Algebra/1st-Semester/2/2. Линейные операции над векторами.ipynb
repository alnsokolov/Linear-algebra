# 2. Линейные операции над векторами. Координаты вектора

## Линейные операции над векторами

### Сложение векторов

Сумма векторов $ a $ и $ b $ - вектор, порождённый отрезком, получающимся из начала направленного отрезка, порождающего вектор $ a $ и конца направленрого отрезка, порождающего $ b $ при совмещении их конца и начала соответственно. Это правило сложения называется правилом треугольника. Этот же вектор $ a + b $ является диагональнью параллелограмма, образованного $ a $ и $ b $. Это правило сложения называется правилом параллелограмма.

### Теорема 1

Операция сложения двух векторов обладает следующими свойствами:

1. $ a + b = b + a, \space \forall a, b $ (свойство коммутативности)
2. $ (a + b) + c = a + (b + c), \space \forall a, b, c $ (свойство аасоциативности)
3. $ \exists 0: a + 0 = 0 + a = a, \space \forall a $ (свойство существования нейтрального элеммента). $ 0 $ называют нулевым вектором
4. $ \forall a \space \exists (-a): a + (-a) = 0 $ (свойство существования симметричного элемента). $ (-a) $ называют противоположным ветором

### Доказательство

1 и 2 свойства доказываются построением. Нулевым ветором будет являться нулевой направленный отрезок, противоположным вектору $ a = \overrightarrow{AB} $ будет $ (-a) = \overrightarrow{BA} $

Разностью векторов $ a, b $ называется такой вектор $ x $, что $ x + b = a $. Обозначается как $ x = a - b $

### Теорема 2

$ \forall a, b, \space \exists! a - b $

### Доказательство

В качестве разности $ a - b $ можно взять $ a + (-b) $, т.к. $ b + (a + (-b)) = (b + (-b)) + a = a $. Эта разность едиственна, так как если $ c $ - ещё одна разность, то $ c = c + 0 = (c + b) + (-b) = a + (-b) $

### Умножение вектора на число

Произведение вектора $ a $ на число $ \alpha $ называется такой вектор $ b $, что:
1. $ |b| = |a| \cdot \alpha $
2. $ b \uparrow \uparrow a, \alpha > 0 $ и $ b \uparrow \downarrow a, \alpha < 0 $

### Теорема 3

Операция умножения вектора на число обладает следующими своствами:
1. $ 1 \cdot a = a $
2. $ (\alpha \beta)a = \alpha(\beta a) $
3. $ (\alpha + \beta)a = \alpha a + \beta a $ (свойство дистрибутивности умножения на число относительно сложения чисел)
4. $ \alpha (a + b) = \alpha a + \alpha b $ (свойство дистрибутивности умножения на число относительно сложения векторов)

## Координаты вектора

Пусть $ e_1 $ - базис $ V_1 $, $ a $ - произвольный вектор из $ V_1 $. Отложим эти векторы от одной точки $ O $ прямой $ V_1 $, так что $ e_1 = \overrightarrow{OE}, a = \overrightarrow{OA_1} $. Тогда $ a = xe_1 $, где

$$ x = \begin{cases} |\overrightarrow{OA_1}|/|\overrightarrow{OE}|, &a \uparrow \uparrow e_1 \newline -|\overrightarrow{OA_1}|/|\overrightarrow{OE}|, &a \uparrow \downarrow e_1 \end{cases} $$

Введём на прямой $ V_1 $ направление, совпадающее с направлением базисного вектора $ e_1 $. Тогда получим $$ x = \frac{(OA_1)}{|e_1|} = \frac{(a)}{|e_1|} $$

Ось, положительное направление которой совпадает с направлением вектора $ e $ называется осью, определённоой вектором $ e $.


При большем количестве базисных векторов, все векторы вновь откладываются от точки $ O $. Базисные вектора образуют отрезки $ OE_1, \dots, OE_n $, а вектор образует отрезок $ OA $. $ A_1, \dots, A_n $ - точки проекций на каждую из прямых, образованных соответствующими отрезками. Координаты вектора будет равны $ \left \{ \frac{(OA_1)}{OE_1}, \dots, \frac{(OA_n)}{OE_n} \right \} $
