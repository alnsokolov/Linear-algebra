# 11. Алгебраические линии и поверхности. Инвариантность порядка линии (поверхности)

Пусть $ Oxy $ и $ Oxyz $ - афинные системы координат на плоскости и в пространстве. Уравнение $ F(x, y) = 0 $ и $ F(x, y, z) = 0 $ называется уравнением линии $ L $ на  плоскости (поверхности) в пространстве. в заданной системе координат, если этому уравнению удовлетворяют координаты всех точек линии $ L $ (поверхности $ \pi $) и только они. Очевидно, что два уравнения определяют одну и ту же линию тогда и только тогда, когда они эквивалентны.  

Алгебраическим одночлеерм относительно переменных $ x, y $ ($ x, y, z $) с коэффицентом $ \lambda $ называется выражение вида $ \lambda x^p y^q $ ($ \lambda x^p y^q z^r $), где $ p, q, r \in \mathbb{N} \cap \{0\} $. Если $ \lambda \neq 0 $, то $ p + q $ ($ p + q + r $) называется степенью многочлена. Алгебраическим многочленом относительно $ x, y $ ($ x, y, z $) с вещественными коэффицентами называется конечная сумма алгебраических одночленов. Наибольшая степень одночлена, входящего в многочлен, называется степенью многочлена.  

Линия на плоскости (поверхность в пространстве) называется алгебраической, если в некоторой афинной системе координат $ F(x, y) $ ($ F(x, y, z) $) - алгебраический многочлен от переменных $ x, y $ ($ x, y, z $) с вещественными коэффицентами. Степень многочлена называется порядком линии.

### Теорема об инвариантности порядка

При переходе от одной афинной системы координат к другой, алгебраическая линия (поверхность) остаётся алгебраической и порядок её не меняется.

### Доказательство (для линии)

Пусть на плоскости в афинной системе координат $ Oxy $ определена линия $ L $, заданная уравнением $ F(x, y) $. При переходе к системе координа $ Ox'y' $, уравнение преобразуется в $ F'(x', y') = 0 $. Рассмотрим его подробнее. Для этого в каждый одночлен $ \lambda x^py^q $ многочлена $ F(x, y) $ вместо $ x $ и $ y $ подставим выражения с $ x' $ и $ y' $. $ \lambda x^py^q = (\alpha + c_{11}x + c_{12}y)^p(\beta + c_{21}x + c_{22}y)^q $. В итоге, мы получили алгебраический многочлен от переменных $ x', y' $, степень которого не превосходит $ p + q $. При этом, алгебраический многочлен $ F(x, y) $ преобразуется в алгебраический многочлен $ F'(x', y') $, степень которого $ n' \leqslant n $. Если в рассуждениях поменять ролями системы координат, то мы получим, что $ n \leqslant n' \Rightarrow n = n' $
