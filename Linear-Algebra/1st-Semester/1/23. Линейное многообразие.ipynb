# 23. Линейное многообразие. Геометрические свойства множества решений неоднородной системы линейных алгебраических уравнений. Общее решение

Пусть $ V $ - линейное пространство, $ L $ - некоторое его подпространство, $ x_0 $ - некоторый вектор пространства $ V $. Множество $ H = \{x_0 + x | x \in L\} = x_0 + L $ называется линейным многообращием пространства $ V $, полученным сдвигом подпространства $ L $ на вектор $ x_0 $. Вектор $ x_0 $ называется вектором сдвига, а $ L $ - направляющим подпространством линейного многообразия $ H $. 

Свойства:

1. Вектор сдвига $ x_0 $ принадлежит линейному многообразию, т.к. $ x_0 = x_0 + \theta, \theta \in L $
2. Разность двух векторов линейного многообразия принадлежит направляющему подпространству: $ z_1 = x_0 + x, z_2 = x_0 + y, x, y \in L \Rightarrow z_1 - z_2 = x - y \in L $

### Теорема

Два линейных многообразия $ H_1 = x_1 + L_1, H_2 = x_2 + L_2 $ совпадают тогда и только тогда, когда $ L_1 = L_2 = L $ и $ x_1 - x_2 \in L $

Следствия:

- Вектором сдвига может служить любой вектор линейного многообразия
- Линейное многообразие может быть получено сдвигом единственного направляющего подпространства

Пусть $ Ax = b $ - неоднородная система линейных алгебраических уравнений. Однородная система $ Ax = 0 $, нолученная из изначальной системы заменой свободных членов нулями, называется приведённой однородной системой для изначальной системы. Между решениями неоднородной и однородной систем существует тесная связь:
1. Сумма решений неоднородной и приведённой однородноой систем является решением неоднородной системы.
2. Разность двух решений неоднородной системы является решением приведённой неоднородной системы.

### Теорема

Множество всех решений неоднородной системы является линейным многообразием, полученным сдвигом подпространства решений приведённой однородной системы на частное решение неоднородной сситемы.

### Доказательство

Пусть $ H $ - множество всех решений неоднородной системы, $ c $ - частное его решение, $ L $ - множество всех решений приведённой системы. Покажем, что $ H = c + L $

Действительно, если $ z \in H $, то $ z = c + (z - c) = c + x $, где $ x = z - c $. Так как $ x \in L $ (как разность двух решений системы), то $ z \in c + L $ и, следовательно $ H \subset c + L $. С другой стороны, если $ z \in c + L $, то $ z = c + x, x \in L $. Значит, $ z \in H $ как сумма решений неоднородной и приведённой систем. Следовательно $ c + L \subset H \Rightarrow H = c + L $

### Общее решение неоднородной системы

Итак, найдя частное решение неоднородной системы и прибавляя его к каждому решению приведённой сситемы, можно получить все решения неоднородной системы в общем виде:

$$ z = c  +\alpha_1 e_1 + \dots + \alpha_{n - r} e_{n - r}, \forall \alpha_1, \dots, \alpha_{n - r} \in \mathbb{R} $$  

где $ c $ - частное решение системы, а $ e_1, \dots, e_{n - r} $ - фундаментальная система решений системы. Это представление решения называется общим решением неоднородной системы уравнений через фундаментальнйю систему решений.
