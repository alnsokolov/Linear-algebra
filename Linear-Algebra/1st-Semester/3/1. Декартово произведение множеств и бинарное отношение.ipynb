# 1. Декартово произведение множеств и бинарное отношение. Отношение эквивалентности. Фактор-множество 

Под множеством в математике понимается совокупность объектов, называемых элементами множества. Множество, не содержащее элементов называется пустым и обозначается $ \emptyset$

### Операции над множествами

- Объединение  
  Множество элементов, которые содержатся и в $ X $, и в $ Y $  
  Обозначение: $ X \cup Y $  
- Пересечение  
  Множество элементов, которые содержатся как в $ X $, так  и в $ Y $  
  Обозначение: $ X \cap Y $  
- Разность множеств  
  Разность множеств $ X $ и $ Y $ - множество всех элементов $ X $, не содержащихся в $ Y $  
  Обозначение: $ X \backslash Y $ или $ X - Y $  
- Декартово произведение  
  Декартовым произведением множеств $ X $ и $ Y $ называется множество всех упорядоченных пар $ (x, y) $, где $ x \in X, y \in Y $
  Обозначение: $ X \times Y $. Итак, $ X \times Y = \{ (x, y) | x \in X, y \in Y \} $

Декартово произведение $ X \times X $ называется декартовым квадратом множества $ X $ и обозначается $ X^2 $. Множество всех пар $ (x, x) $, где $ x \in X $ называется диагональю декартова квадрата множества $ X $.

### Эквивалентность

#### Бинарное отношение

Говорят, что на множестве $ X $ задано бинарное отношение $ R $, если указано непустое множество $ R $ декартова квадрата этогомножества. Если при этом $ (x, y) \in R $, то говорят, что $ x $ и $ y $ связаны отношением $ R $, и обозначают символом $ xRy $.

Равенство $ x = y $ и неравенство $ x < y $ действительных чисел, равенство матриц $ A = B $, равенство множеств $ X = Y $ и т.п. - это примеры бинарных отношений.

Бинарное отношение $ R $ на множестве $ X $ называется:
- рефлексивным, если $ xRx \space\space \forall x\in X $
- симметричным, если имеет место импликация $ xRy \Rightarrow yRx $
- транзитивным, если имеет место импликация $ xRy, yRz \Rightarrow xRz $

#### Отношение эквивалентности

Бинарное отношение $ E $ на множестве $ X $ называется отношением эквивалентности, если оно рефлекторно, симметрично и транзитивно.Если пара элеентов $ x, y \in X $ связана отношением эквивалентности, то говорят, что $ x $ и $ y $ эквивалентны и обозначают это как $ x \sim y $. В конкретных случаях могут быть использованы тругие символы, такие как $ x = y $ и $ x \equiv y $

Пусть на множестве $ X $ задано отношение эквивалентности $ E $ и $ a \in X $. Множество всевозможныхэлементов $ x \in X $, эквивалентных $ a $ называется классом эквивалентности, порождённм элементом $ a $. Обозначение: $ cl(a) $. Итак $ cl(a) = \{ x \in X | x \sim a \} $. Любой элемент класса эквивалентности называется представителем класса.

### Теорема 1

Класс эквивалентности порождается любым своим представителем, т.е. если $ b \in cl(a) $, то $ cl(b) = cl(a) $

### Доказательство

Достаточно показать двусторонне вложение $ cl(a) $ и $ cl(b) $:

$ x \in cl(a) \Rightarrow x \sim a $, но $ a \sim b \Rightarrow c \sim b \Rightarrow c \in cl(b) $  
$ x \in cl(b) \Rightarrow x \sim b $, но $ b \sim a \Rightarrow c \sim a \Rightarrow c \in cl(a) $

### Теорема 2

Два класса эквивалентности либо не пересекаются, либо совпадают.

### Доказательство

Действительно, два класса эквивалентности либо не пересекаются, либо имеют хотя бы один общий элемент.Но в силу теоремы 1, в этом сулчае они совпадают.

Из теоремы 2 и очевидного факта, что любой элемент множества содержится в одном из классов эквивалентности, следует, что всё множество разбивается на неперсекающиеся классы эквивалентности. Множество всех классов эквивалентности множества $ X $ называется фактор-множеством $ X $ по отношению эквивалентности $ E $ и обозначается $ X|E $.
