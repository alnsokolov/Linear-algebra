# 17. Тригонометрическая форма комплексного числа. Модуль и аргумент произведения комплексных чисел

Положение комплексного числа на комплексной плоскости также может быть представлено и с помощью её полярных координат. В соответствии с этим вводятся следующие характеристики комплексного числа.

Модулем комплексного числа $ z = a + bi $ называется число $ r = \sqrt{a^2 + b^2} $. Обозначение: $ |z| $. Из определения вытекают следующие свойства:
1. $ |z| \in \mathbb{R}, |z| \geqslant 0 $, причём $ |z| = 0 \Leftrightarrow z = 0 $
2. $ |z| $ совпадает с полярным радиусом точки, изображающей комплексное число на комплексной плоскости
3. $ |z| = \sqrt{z\overline{z}} $
4. Модуль числа совпадает с абсолютной величиной этого числа

Аргументом комплексного числа называется угол $ \varphi $ между положительным направлением оси абсцисс и радиус-вектором точки $ M $, отсчитываемый от оси абсцисс в любом направлении, при этом положительным считается направление против часовой стрелки. Обозначение $ arg \space z $. Из определения вытекают свойства:
1. $ arg \space z $ не определён для $ z = 0 $, а для $ z \neq 0 $ определён с точностью до $ 2\pi $
2. $ arg \space z $ отличается от полярного угла точки $ M $ тем, что он имеет бесконечно много значений
3. Два комплексных числа равны тогда и только тогда, когда $$ \begin{cases} |z_1| = |z_1|, и \space\space если \space\space |z_1| \neq 0 \newline arg \space z_1 = arg \space z_2 + 2\pi k, k \in \mathbb{Z} \end{cases} $$

### Теорема 1

Любое комплексное число может быть записано в виде $$ z = r (cos \space \varphi + i \space sin \space \varphi) $$ где $ r = |z|, \varphi = arg \space z $

### Доказательство

$ z = a + bi \Rightarrow a = r \space cos \space \varphi, b = r \space sin \space \varphi \Rightarrow z = r (cos \space \varphi + i \space sin \space \varphi) $

Данная форма записи комплексного числа называется тригонометрической формой этого числа.

### Теорема 2

При умножении комплексных чисел, их модули умножаются, а аргументы складываются. При делениии комплексных чисел, их модули делятся, а аргументы вычитаются.

### Доказательство

Доказывается умножением в тригинометрической форме.
