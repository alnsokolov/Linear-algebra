# Вопрос 4

ОДУ в симметричном виде. Понятие параметрического решения и общего интеграла (ОИ). Уравнения в полных дифференциалах (УПД), существование ОИ. Теорема о необходимом и достаточном условии УПД.

## Ответ

Дифференциальным уравнением в симметричном виде (или в дифференциалах) называется уравнение
$$ M(t, y)dt + N(t, y)dy = 0 $$

Вредпологается, что $ M $ и $ N $ определены и непрерывны в некоторой области $ D \subseteq \mathbb R^2 $ и
$$ |M(t, y)| + |N(t, y)| > 0, \quad \forall (t, y) \in D $$
Это уравнение является более общим посравнению с уравнением $ y'(t) = f(t, y(t)) $, поскольку последнее можно представить в виде этого при $ N(t, y) = -1, M(t, y) = f(t, y) $

Пара функций $ t = \phi(\tau), y = \psi(\tau) $ называется параметрическим решением уравнения в симметричком виде $ M(t, y)dt + N(t, y)dy = 0 $ на отрезке $ [\tau_1, \tau_2] $ если:

1. $ (\phi(\tau), \psi(\tau)) \in C^1[\tau_1, \tau_2] $ и $ |\phi(\tau)| + |\psi(\tau)| > 0 \quad \forall \tau \in [\tau_1, \tau_2] $
2. $ (\phi(\tau), \psi(\tau)) \in D, \; \forall \forall \tau \in [\tau_1, \tau_2] $
3. $ M(\phi(\tau), \psi(\tau)) \phi'(\tau) + N(\phi(\tau), \psi(\tau)) \psi'(\tau) = 0, \quad \forall \tau \in [\tau_1, \tau_2] $

Уравнение $ \Phi (t, y, c) = 0 $ называется интегралом дифференциального уравнения в области $ D $, если при любом $ c \in C_0 $ оно определяет решение уравнения.

Интеграл наpывается общим, если он определяет все решения уравнения, то есть для любого решения уравнения $ t = \phi(\tau), y = \psi(\tau) $, интегральная кривая которого лежит в $ D $, найдётся постоянная $ \overline c \in C_0 $ такая, что $ \Phi(\phi(\tau), \psi(\tau), \overline c) \equiv 0 $

Так как общий интеграл определяет все решения дифференциального уравнения, то в том случае, когда его удается найти, задача поиска решений дифференциального уравнения считается решенной.

Дифференциальное уравнение в симметричном виде $ M(t, y)dt + N(t, y)dy = 0 $ называется уравнением в полных дифференциалах в области $D$, если существует непрерывно дифференцируемая в $D$ функция $V (t, y)$ такая, что $ \displaystyle|\frac{\partial V(t, y)}{\partial t}| + |\frac{\partial V(t, y)}{\partial y}| > 0 $ и

$$ M(t, y) = \frac{\partial V(t, y)}{\partial t}, N(t, y) = \frac{\partial V(t, y)}{\partial y}, \quad \forall (t, y) \in D $$

## Теорема

Уравнение в полных дифференциалах вида $ M(t, y)dt + N(t, y)dy = 0 $ имеевт в области $ D $ общий интеграл

$$ V(t, y) = c $$

## Доказательство

Рассматриваем это уравнение вблизи окарестности произвольной точки $ (t_0, y_0), V(t_0, y_0) = c_0 $. Из условий представления уравнения $ M(t, y) = \frac{\partial V(t, y)}{\partial t} \neq 0 $ или $ N(t, y) = \frac{\partial V(t, y)}{\partial y} \neq 0 $. Положим, что второе верно. Тогда по теореме о неявной функции в окрестности рассматриваемой точки $ \exists! g(t) = y: g(t) \in C^1, y_0 = g(t_0) $ и $ V(t, g(t)) = c_0 $. Берём дифференциалы правой и левой частей уравнения. Профит, это решение.

Для доказания того, что это общий интеграл, показываем, что $ \exists c: V(\phi(\tau), \psi(\tau)) = c, \quad \forall \tau \in [\tau_1, \tau_2] $. Дифференцируем $ V(\phi(\tau), \psi(\tau)) $, имеем его равным 0 в силу представления $ M $ и $ N $. Значит $  V(\phi(\tau), \psi(\tau)) = c $, профит

## Теорема

Пусть функции $ M (t, y), N (t, y) $ и их частные производные первого порядка непрерывны в прямоугольнике $D$ со сторонами, параллельными координатным осям, и выполнено условие $ |M(t, y)| + |N(t, y)| > 0 $. Тогда для того, чтобы уравнение $ M(t, y)dt + N(t, y)dy = 0 $ было уравнением в полных дифференциалах в $ D $, необходимо и достаточно, чтобы
$$ \frac{\partial M(t, y)}{\partial y} = \frac{\partial N(t, y)}{\partial t} $$

## Доказательство

Необходимость. Говорим, что существует общий интеграл $ V(t, y) $, дифференцируем его два раза по обоим переменным с изменением порядка дифференцирования. Профит, получаем требуемое равенство.

Достаточность. Рассмотрим 

$$ V(t, y) = \displaystyle\int_{t_0}^t M(\xi, y) d \xi + \int_{y_0}^y M(t_0, \eta) d \eta $$

Дифференцированием по $ t $ и $ y $ получаем $ M(t, y) = \frac{\partial V(t, y)}{\partial t}, N(t, y) = \frac{\partial V(t, y)}{\partial y} $. Профит
