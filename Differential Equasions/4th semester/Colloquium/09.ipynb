# Вопрос 9

Особые решения уравнения первого порядка. Необходимые условия, примеры.

## Ответ

Функция $ y = \xi(t) $ называется особым решением дифференциального уравнения

$$
F(t, y(t), y'(t)) = 0
$$

на отрезке $ [t_1, t_2] $ если $ y = \xi(t) $ является обычным решением уравнения на этом отрезке, и через каждую точку интегральной кривой $ \Gamma = \{ (t, y): \space y = \xi(t), \space t \in [t_1, t_2] \} $ этого решения проходит другое решение этого уравнения с тем же самым наклоном касательной, но отличающееся от данного в сколь угодно малой окрестности точки.

Т.о. в каждой точке интегральной кривой особого решения нарушается едиснтвенность решения задачи Коши

$$ F(t, y(t), y'(t)) = 0, \quad y(t_0), = y_0, \quad y'(t_0) = y_0', \quad \forall(t_0, y_0) \in \Gamma $$

Значит нарушается одно или несколько условий теоремы о существовании и единственности решения задачи Коши.

Например, особое решение может содержаться среди тех кривых, на которых не существуют производные по $ p, y $. Пусть это не так. Тогда во всех точках интегральной кривой особого решения должны выполняться равенства 

$$
F(t, \xi(t), \xi'(t)) = 0, \quad \frac{\partial F (t, \xi(t), \xi'(t))}{\partial p} = 0
$$

То есть тройка $ (t, \xi(t), \xi'(t)) $ при каждом $ t $ должна являться решением системы

$$
\begin{cases} F(t, y, p) = 0 \\ \displaystyle\frac{\partial F}{\partial p}(t, y, p) = 0 \end{cases}
$$

Часто из этой системы можно исключить переменную $ p $ и получить уравнение $ \Phi(t, y) = 0 $. Решения этого уравнения на плоскости задаются одной или несколькими линиями и называются дискриминантными кривыми.

Возможны 3 случая:
1. Уравнение $ \Phi(t, y) = 0 $ задаёт особое решение
2. Уравнение $ \Phi(t, y) = 0 $ задаёт решение $ F $, но оно не особое
3. Уравнение $ \Phi(t, y) = 0 $ задаёт функцию, не являющуюся решением

### Примеры

1. $ (y')^3 - y^2 = 0 $
2. $ (y')^2 - y^2 = 0 $
3. $ (y'(t))^2 - (t + y(t)) y'(t) + t y(t) = 0 $
