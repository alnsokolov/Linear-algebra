# 15. Ранг произведения матриц. Ранг матрицы и элементарные преобразования

### Лемма

Если все строки (столбцы) матрицы $ A $ линейно выражаются через строки (столбцы) матрицы $ B $, то $ rg \ A \leqslant rg \ B $

### Доказательство

Пусть $ rg \ A = r, rg \ B = s $ и $ a_1, \dots, a_r $ и $ b_1, \dots, b_s $ - базисные столбцы матриц. Докажем, что $ r \leqslant s $.  

Допустим $ r > s $. Тогда согласно условию теоремы система $ a_1, \dots, a_r $ выражается через систему столбцов матрицы $ B $, которая, в свою очередь выражается через сситему базисных столбцов $ b_1, \dots, b_s $. Значит $ a_1, \dots, a_r $ выражается через $ b_1, \dots, b_s $, причём $ r > s $. Из теоремы о двух системах, система $ a_1, \dots, a_r $ линейно зависима. Это противоречит тому, что $ a_1, \dots, a_r $ - базисные столбцы.

### Ранг произведения матриц

Ранг произведения матриц не превосходит рангов сомножителей

### Доказательство

Пусть $ C = AB $. Из свойств произведения матриц следует, что строки матрицы $ C $ линейно выражаются через строки матрицы $ B $, а столбцы - через столбцы матрицы $ A $. В силу леммы выше $ rg \ C \leqslant rg \ A, \ rg \ C \leqslant rg \ B $

### Инвариантность ранга относительно элементарных преобразований

Ранг матрицы не изменяется при умножении её на матрицу элементарных преоразований.

### Доказательство

Все матрицы элементарных преобразований являются невырожденными. Докажем, что при умножении матрицы на невырожденную матрицу, её ранг не меняется.

Пусть $ C = AP, |P| \neq 0 $. Тогда на основании теоремы о ранге произведения имеем $ rg \ C \leqslant rg \ A $. С другой стороны, матрица $ P $ обратима $ \Rightarrow A = CP^{-1} \Rightarrow rg \ A \leqslant rg \ C \Rightarrow A = C $. Отсюда вытекает и то, что $ rg \ QA = rg \ A, |Q| \neq 0 $
