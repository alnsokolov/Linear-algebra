# 2. Приведение матрицы к ступенчатому виду. Приведение к диагональному виду.

### Элементарные преобразования

1. Перестановка двух строк (столбцов) матрицы
2. Умножение строки (столбца) матрицы на число, отличное от нуля
3. Прибавление к одной строке (столбцу) матрицы другой её строки (столбца), умноженной на любое число

### Теорема (об основном процессе)

Произвольная ненулевая матрица конечным числом элементарных преобразований только строк первого и третьего типов,, может быть приведена к верхней ступенчатой форме.  

$ A = (a_{ij}) \in \mathbb{R}, A \neq 0 $

#### Доказательство

Т.к. $ A \neq 0 $ существует ненулевой столбец. Пусть $ k_1 $ - номер первого из них. Если $ a_{1k_1} \neq 0 $, то переходим дальше, иначе меняем строки 1 и $ i $, где $ i $ - номер первой строки в этом столбце с ненулевым элементом.  

Элемент $ a_{1k_1} $ назовём ведущим и обнулим с его помощью все элементы ниже. Для этого из каждой строки вычтем первую, умноженную на $ \frac{a_{2k_1}}{a_{1k_1}}, \frac{a_{3k_1}}{a_{1k_1}}, \dots, \frac{a_{nk_1}}{a_{1k_1}} $. После этого шага первая строка станет первой строкой строящейся ступенчатой матрицы. Если все стролбцы ниже $ 1 $ нулевые, то матрица приведена к треугольному виду. Иначе, применим вышеописанный процесс к матрице, образованной столбцами $ k_1 + 1, k_1 + 2, \dots $ и строками $ 2, 3, \dots $. Повторяя эти преобразования не более чем через $ min(n, m) $ шагов мы получим матрицу ступенчатого вида.

При помощи применения основоного опроцесса сначала к строкам, а затем к столбцам квадратной матрицы, можно получить диагональную матрицу.
