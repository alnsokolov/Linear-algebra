# 3. Перестановки

Упорядоченная совокупность чисел $ \{\alpha_1, \alpha_2, \dots, \alpha_n\} $ называется перестановкой, если:

1. $ \alpha_i \in {1, 2, \dots, n}, i = \overline{1, n} $
2. $ \alpha_i \neq \alpha_j $ при $ i \neq j $

Последовательность $ 1, 2, \dots, n $ называется натуральной перестановкой.

Преобразование перестановки, при котором менябтся местами два числа, называется транспозицией.  

Говорят, что числа $ a_i, a_j $ образуют в перестановке инверсию (беспорядок), если $ i < j, a_i  > a_j $.  

Перестановка называется чётной, если количество инверсий в ней чётно, а иначе - нечётной. Число инверсий в перестановке обозначается $ \sigma(\alpha_1, \alpha_2, \dots, \alpha_n) $ или $ \sigma(\alpha) $

### Теоремы

- Число всевозможныых перестановок из $ n $ чисел равно $ n! $
- Каждая транспозиция меняет чётность перестановки
- Все $ n! $ перестановой могут быть упорядочены так, чтобы предыдущая отличалась от последующей на одну транспозиций, причём начинать это упорядочение можно с любой перестановки
- Если $ \alpha_1, \alpha_2, \dots, \alpha_n $ - перестановка из первых $ n $ натуральных чисел с числом инверсий $ s $ то после преобразования её в натуральную перестановку индексные номера $ 1, 2, \dots, n $ образуют новую перестановку с числом инверсий $ s $

### Следствия

- При $ n \geqslant 2 $ число нечётных перстановок равно числу чётных
- От каждой из перстановок из $ n $ чисел можно перейти к любой другой перестановке из этих же чисел с помощью конечного числа транспозиций
