# 4. Определитель. Свойства определителя

Определителем матрицы $ A = (a_{ij}) \in \mathbb{R}^{n \times n} $ называется сумма всевозможных произведений $ a_{1 \alpha_1} a_{ \alpha_2} \dots a_{n \alpha_n} $ элементов матрицы, взятых по одному из каждой строки, причём если сомножители упорядочены в порядке возрастания строк, то произведение взято со знаком $ (-1)^{\sigma(\alpha)} $

Обозначается он как $ |A| = \sum\limits_{\alpha = (\alpha_1, \alpha_2, \dots, \alpha_n)} (-1)^{\sigma(\alpha)} a_{1 \alpha_1} a_{2 \alpha_2} \dots a_{n \alpha_n} $, где суммирование ведётся по всем возможным перестановкам $ (\alpha_1, \alpha_2, \dots, \alpha_n) $ из чисел $ 1, 2, \dots, n $

### Свойства определителя

1. Определитель треугольной матриц равен произведению элементов её диагонаи
2. Определитель квадратной матрицы не изменяется при её транспонировании: $ |A| = |A^T| $
    - Следствие: в определение определителя можно поменять местами строки и столбцы: $ |A| = \sum\limits_{\alpha = (\alpha_1, \alpha_2, \dots, \alpha_n)} (-1)^{\sigma(\alpha)} a_{\alpha_1 1} a_{\alpha_2 2} \dots a_{\alpha_n n} $
3. Если одна из строк (столбцов), является нулевым, то определитель равен нулю
4. При умножении строки (столбца) матрицы на число, её определитель умножется на это же число
5. Если каждый элемент строки матрицы представим в виде суммы двух слагаемых - $ a_{ik} = b_k + c_k, k = \overline{1, n} $, - то определитель такой матрицы можно представить в виде суммы двух определителей:  
$
\begin{vmatrix}
a_1' \\
\vdots \\
b' + c' \\
\vdots \\
a_n'
\end{vmatrix}
=
\begin{vmatrix}
a_1' \\
\vdots \\
b' \\
\vdots \\
a_n'
\end{vmatrix}
+
\begin{vmatrix}
a_1' \\
\vdots \\
c' \\
\vdots \\
a_n'
\end{vmatrix}
, b' = (b_1, b_2, \dots, b_n), c' = (c_1, c_2, \dots, c_n)
$  

6. При перетановке двух строк (столбцов) матрицы, определитель меняет знак
7. Определитель матрицы, имеющей две одинаковые строки (столюца) равен нулю
8. Если одна из строк (столбцов) матрицы является линейноу комбинацией других её строк (столбцов), то опеределитель равен нулю
9. Если к какой-либо строке (столбцу) прибавить линейную комбинацию других её строк (столбцов), то её определитель не изменится.
