# 22. Линейное подпространство. Геометрические свойства множества решений однородной системы линейных алгебраических уравнений. Фундаментальная система решений

Непустое подмножжество $ L $ пространства $ V $ называется линейным подпространством $ V $, если оно само является линейным пространством относительно законов композиции, действующим в $ V $.

Это возможно тогда и только тогда, когда имеют место импликации  

$$
\begin{array}
\ a, b \in L                     &\Rightarrow& a + b \in L \\
a \in L, \alpha \in \mathbb{R} &\Rightarrow& \alpha a \in L
\end{array}
$$  

Эти свойства называются свойствами замкнутости множества $ L $ относительно указанных операций.

### Линейное пространство решений однородной системы

Множество всех решений однородной системы $ Ax = 0 $ с $ n $ неизвестными является линейным подпространством арифметического пространства $ \mathbb{R}^n $.

### Доказательство

Обозначим $ L = \{x \in \mathbb{R}^n | Ax = 0\} $. Легко проверить, что если $ x \in L, y \in L $, то $ x + y \in L $ и, кроме того, если $ \alpha \in \mathbb{R}, x \in L $, то $ \alpha a \in L $. Следовательно, $ L $ - линейное подпространство пространства $ \mathbb{R}^n $.

Произваольный базис подпространства решений однородной системы линейных алгебраических уравнений называется фундаментальной системой решений. Она существует лишь в том случае, если существует нетривиальное решение системы уравнений. При этом система может обладать многими фундаментальными решениями.

### Теорема

Размерность пространства решениц однородной системы $ Ax = 0 $ с $ n $ неизвестными равна $ n - r $, где $ r = rg \ A $.

### Доказательство

Псотроим фундаментальную систему решений. Для этого придадим свободным неизвестным $ x_{r + 1}, \dots, x_n $ следующие $ n - r $ наборов решений: $ (1, 0, \dots, 0), (0, 1, \dots, 0), \dots, (0, 0, \dots, 1) $. Для каждого из этих наборов найдём соответствующее значения главных неизвестных. Тем самым найдём $ n - r $ решений системы:  

$$
\begin{array}
&e_1 &=& (x_{11}, &c_{12}, &\dots, &c_{1r}, &1, &0, &\dots, &0)^T, \\
e_2 &=& (x_{21}, &c_{22}, &\dots, &c_{2r}, &0, &1, &\dots, &0)^T, \\
\cdots \\
e_{n - r} &=& (x_{n-r,1}, &c_{n-r,2}, &\dots, &c_{n-r,r}, &0, &0, &\dots, &1)^T
\end{array}
$$

Свойства решений $ e_1, e_2, \dots, e_{n - r} $:
1. Решения $ e_1, \dots, e_{n - r} $ линейно незавилимы, т.к. матрица  
$$
C = \begin{bmatrix} e_1 \\ e_2 \\ \vdots \\ e_{n - 1} \end{bmatrix}
$$

содержит ненулевой минор порядка $ n - r $ справа.
2. Любое решение $ x = (x_1, x_2, \dots, x_r, x_{r+1}, \dots, x_n)^T $ является линейной комбинацией решений $ e_1, e_2, \dots, e_{n - r} $ и, более того, $ x = x_{r + 1}e_1 + x_{r + 2}e_2 + \dots + x_n e_{n - r} $

#### Доказательство

Действительно, обозначим $ y = x - x_{r + 1}e_1 - \dots - x_n e_{n - r} $. Пусть $ y = (y_1, \dots, y_n)^T $. Очевидно, что $ y $ - решение системы. Найдём его, пользуясь общей теорией решения систем: придадим свободным неизвестным значения $ y_1, \dots, y_n $. Легко продемонстрировать, что $ y_1 = y_2 = \dots = y_n = 0 $. Следовательно, укороченная система будет однородной системой, имеющей единственное и тривиальное решение. Значит $ y = 0 \Rightarrow x = x_{r + 1}e_1 - \dots - x_n e_{n - r} \ $ и $ \ e_1, \dots, e_{n - r} $ - фундаментальная система решений.

Построенная фундаментальная система решений называется нормальной фундаментальной системой решений.
