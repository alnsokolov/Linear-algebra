# 10. Линейная зависимость в линейном пространстве

Пусть $ \alpha_1, \dots, \alpha_k \in \mathbb{R}, \ \ a_1, \dots, a_k \in V $, где $ V $ - линейное пространство. Вектор $ \alpha_1 a_1 + \dots + \alpha_k a_k $ называется линейной комбинацией векторов $ a_1, \dots, a_k $ с коэффицентами $ \alpha_1, \dots, \alpha_k $. Если вектор $ b $ является линейной комбинацией $ a_1, \dots, a_k $, то говорят, что он линейно выражается через векторы $ a_1, \dots, a_k $, при этом предстваление $ b = \alpha_1 a_1 + \dots + \alpha_k a_k $ называется линейным разложением вектора $ b $ по векторам $ a_1, \dots, a_k $. Очевидно, что нулевой вектор выражается линейно через любой вектор, т.к. $ \theta = 0a, \forall a \in V $. Через нулевой вектор не может выражаться ни один ненулевой вектор, т.к. $ \alpha \theta = \theta, \forall \alpha \in \mathbb{R} $.

Линейная комбинация называется тривиальной, если все её коэффиценты равны нулю, в ином случае она называется нетривиальной. Очевидно, что тривиальная комбинация любых векторов равна нулевому вектору.

Система векторов называется линейно зависимой, если  

$$ \exists \alpha_1, \alpha_2, \dots, \alpha_k, \exists \alpha_i \neq 0: \alpha_1 a_1 + \alpha_2 a_2 + \dots + \alpha_k a_k = \theta $$  

и линейно независимой, если лишь тривиальная комбинация векторов равна нулевому вектору.

#### Теорема

Система из одного вектора линейно зависима тогда и только тогда, когда этот ветор - нулевой

#### Теорема

Система из векторов $ a_1, a_2, \dots, a_k $, где $ k > 1 $ линейно зависима тогда и только тогда, когда хотя бы один из её векторов линейно выражается через другие.

#### Теорема

Если вся подсистема системы векторов линейно зависима, то и вся система линейно зависима.

#### Теорема

Система векторов линейно независима тогда и только тогда, когда любой вектор, являющийся линейной комбинацией этих векторов, имеет единственное разложение по этим векторам.

#### Теорема

Если система векторов $ a_1, a_2, \dots, a_k $ линейно независима, а система $ a_1, a_2, \dots, a_k, b $ линейно зависима, то ветор $ b $ линейно выражается через вектора $ a_1, a_2, \dots, a_k $.
