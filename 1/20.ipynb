# 20. Исследование и решение системы линейных алгебраических уравнений общего вида. Главные и свободные неизвестные. Общее решение

Не нарушая общности рассуждений, бцдем считать, что базисный минор матрицы $ A $ расположен в её верхнем левом углу. Рассмотрим укороченную систему из $ r $ из н первых уравнений:

$$
\begin{cases}
a_{11}x_1 &+ \dots &+ a_{1r}x_r &+ a_{1, r + 1}x_{r + 1} &+ \dots &+ a_{1n}x_n = b_1 \\
\cdots \\
a_{r1}x_1 &+ \dots &+ a_{rr}x_r &+ a_{r, r + 1}x_{r + 1} &+ \dots &+ a_{rn}x_n = b_r
\end{cases}
(**)
$$  

Укороченная система эквивалентна исходной системе. Очевидно, что любое решение старой системы является решением новой, а, так как все строки старой расширенной матрицы являются линейной комбинацией первых $ r $ строк, любое решение новой матрицы обращает в тождества все последующие уравнения.

Если $ r = n $, система имеет единственное решение как система с невырожденной квадратной матрицей.  

Пусть $ r < n $. Назовём неизвестные, входящие в базисный минор главными, а остальные - свободными. Придадим свободным неизвесным произвольные значения $ c_{r+1}, c_{r + 2}, \dots, c_{n} $ и перенесём их в правую часть системы. В итоге, получается система с квадратной невырожденной матрицей. Эта система имеет единственное решение $ c_1, \dots, c_r $. Очевидно, что совокупность $ (c_1, \dots, c_r, c_{r + 1}, \dots, c_n) $ является решением системы.

Придавая свободным неизвестным произвольные значения и вычисляя главные неизвестные, можно получить любое решение системы.

Система линейных алгебраических уравнений с $ n $ неизвестными имеет единствееное решение тогда и только тогда, когда $ rg \ A = rg \ B = n $

### Общее решение

В случае определённой системы это означает найти определённое решение, в случае неопределённой - описать множество её решений. Один из таких способов состоит в следующем: решим системум относительно главных неизвестных:

$$
x_1 = f_1 (x_{r + 1}, \dots, x_{n}), \\
\cdots \\
x_r = f_r (x_{r + 1}, \dots, x_{n})
$$  

Где $ f_1, f_r $ - некоторые однозначно определяемые функции. Соотношения при произвольных $ x_{r + 1}, \dots, x_n $ описывают множество всех решений сситемы и называются общим решением системы.
