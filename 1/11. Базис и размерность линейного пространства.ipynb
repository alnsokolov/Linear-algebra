# 11. Базис и размерность линейного пространства

Базисом линейного пространства называется упорядоченная линейно независимая система векторов пространства, через которую линейно выражается любой другой вектор пространства. Согласно этому правилу, система векторов $ e_1, \dots, e_n \in V $ образует базис линейного пространства $ V $, если:
- Система $ e_1, \dots, e_n $ линейно независима
- $ \forall x \in V \exists x_1, \dots, x_n: x = x_1 e_1 + \dots + x_n e_n $

### Теорема

Система $ e_1, \dots, e_n $ является базисом линейного пространства тогда и только тогда, когда она обрузует максимальную линейно независимую систему векторов этого пространства.

### Доказательство

#### Необходимость

Базис линейного пространства является максимальной линейно независимой системаой векторов в этом пространстве, так как любая большая система векторов линейно выражается через этот базис и, следственно, линейно зависима.

#### Достаточность

Пусть $ e_1, \dots, e_n $ - максимальная линейно независимая система векторов пространства $ V $, тогда как для любого вектора $ x \in V $ система векторов $ e_1, \dots, e_n, x $ линейно зависима, так как содержит более чем $ n $ веторов. Значит вектор $ x $ является линейной комбинацией $ e_1, \dots, e_n $. Следовательно, векторы $ e_1, \dots, e_n $ образуют базис пространства $ V $.

Количество векторов в базисе называется размерностью пространства $ V $. Обозначается $ dim \ V $  

Линейное пространство размерности $ n $, где $ n $ - целое неотрицательное число, называется $ n $-мерным пространством. Любое $ n $-мерное пространство называется конечномерным линейным пространством. Линейное пространство называется бесконечномерным, если $ \forall k \in \mathbb{N} $ существует линейно независимая система из $ k $ векторов.
