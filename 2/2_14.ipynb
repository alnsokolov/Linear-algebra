# 14. Общее уравнение плоскости в пространстве в афинной системе координат. Критерий параллельности вектора плоскости

### Теорема 1

Поверхность в пространстве является плоскостью тогда и только тогда, когда она представляет собой алгебраическую поверхность первого порядка.

### Доказательство. Необходимость

Пусть $ \pi $ - плоскость в пространстве, проходящая череез точку $ M_0 $ и параллельная ненулевым векторам $ p_1, p_2 $. Пусть $ Oxyz $ - произвольная афинная система координат и $ p_1 = \{ m_1, n_1, k_1 \}, p_2 = \{ m_2, n_2, k_2 \}, M_0(x_0, y_0, z_0) $. Тогда плоскость описывается каноническим уравнением, или, то же самое, уравнением $ \space (x - x_0)(n_1k_2 - n_2k_1) - (y - y_0)(m_1k_2 - m_2k_1) + (z - z_0)(m_1n_2 - m_2n_1) = 0$, которое, если положить $ A = n_1k_2 - n_2k_1, B = -m_1k_2 + m_2k_1, C = m_1n_2 - m_2n_1, D = -x_0 \cdot A + y_0 \cdot B - z_0 \cdot C $, может быть записано в виде  

$$ Ax + By + Cz + D = 0 $$  

Так как вектора $ p_1 \neq 0, p_2 \neq 0 $ и они не коллинеарны, то по крайней мере один из коэффицентов $ A, B $ или $ C $ отличен от нуля. Поэтому левая часть уравнения представляет собой алгебраический многочлен первой степени. Следовательно, любая плоскость в пространстве является алгеюраической поверхностью первого порядка.

### Достаточность

Пусть в афинной системе координат $ Oxyz $ плоскость $ \pi $ определяется уравнением $ Ax + By + Cz + D = 0 $. Это уравнение умеет частное решение $ x_0 = -\frac{AD}{A^2 + B^2 + C^2}, y_0 = -\frac{BD}{A^2 + B^2 + C^2}, z_0 = -\frac{CD}{A^2 + B^2 + C^2} $, ибо $ Ax_0 + By_0 + Cz_0 + D = 0 $. Вычитая последнее равенство из общего уравнения прямой, получим, что $ A(x - x_0) + B(y - y_0) + C(z - z_0) = 0 $, что то же самое,  
$$ \left | \begin{align} x - x_0 && y - y_0 && z - z_0 \newline -B && A && 0 \newline -C && 0 && A \end{align} \right | = 0 $$  

Это уравнение определяет плоскость, проходящую через $ M_0{x_0, y_0, z_0} $ с вектором нормали $ n = \{ A, B, C \} $

### Теорема 2

В афинной системе координат $ Oxyz $ в пространстве вектор $ a = \{ m, n, k \} $ параллелен плоскости, заданной общий уравнением тогда и только тогда, когда $ Am + Bn + Ck = 0 $

### Доказательство

Как следует из доказательства теоремы 1 вектор $ n = \{A, B, C\} $ является вектором нормали данной прямой. Это означает, что вектор $ a $ параллелен этой прямой огда и только тогда, когда $ a $ перпендекулярен $ n $ или $ (a, n) = 0 $. В координатах скалярное умножение векторов выглядит как $ Am + Bn + Ck = 0 $.
