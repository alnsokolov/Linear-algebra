# 25. Гипербола. Фокусы и дирректрисы

Гиперболой называется геометрическое место точек $ M $ плоскости, для которых абсолютная величина разности двух расстояний до двух фиксированных точек $ F_1, F_2 $ есть постоянное полодительное число, меньшее, чем расстояние между $ F_1 $ и $ F_2 $. Это число обозначим за $ 2a $. Точки $ F_1, F_2 $ называются фокусами гмперболы, расстояние между ними называется фокальным расстоянием и обозначается через $ 2c $. Числа $ r_1 = \rho(M, F_1), r_2 = \rho(M, F_2) $ называются фокальными радиусами точки $ M $. Из определния следует, что точка $ M $ является точкой гиперболы тогда и только тогда, когда $ |r_1 - r_2| = 2a, \space\space a < c $

Как и для эллипса построим каноническую систему координат $ Oxy $ для гиперболы.

### Теорема 1

Уравнение гиперболы в канонической системе координатимеет вид $$ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $$

### Доказательство

Аналогично тому, что для эллипса.

Кроме того, $ a^2 = c^2 - b^2 $ и $ r_i = \frac{cx}{a} + (-1)^ia, \space\space\space\space i = 1, 2 $

### Свойства гиперболы

1. Координатные оси являются осями симметриии гиперболы, начало координат - центром симметрии шиперболы. Ось $ Ox $ пересекает гиперболу в точках $ A_1(-a, 0), A_2(a, 0) $ и называется вещественной осью гиперболы, ось $ Oy $ не пересекает её и называется мнимой осью гиперболы. Числа $ a > 0, b > 0 $ называются вещественной и мнимой полуосями гиперболы.
2. Все точки гиперболы лежат внне полосы, определяемой прямыми $ x = \pm a $. Это значит, что она состоит из двух частей, называемых ветвями гиперболы.
3. Все точки гиперболы лежат в тех вертикальных углах, образованных прямыми $ y = \pm \frac{b}{a}x $, которые содержат действительную ось
3. Исследуем форму гиперболы в первой четверти координатной системы. Пусть $ y(x) = \frac{b}{a}\sqrt{x^2 - a^2} $
   - $ y(x) $ непрерывна и возрастает на $ [a, +\infty) $, при этом $ f(a) = 0, \lim_{x \to +\infty} \limits y(x) = +\infty  $
   - $ y'(a) = \infty \Rightarrow $ касательная к гиперболе в точке $ x = a $ параллельна оси $ Oy $
   - $ y''(x) < 0 \Rightarrow $ линия выпукла вверх
   - Если $ l $ - прямая, заданная уравнением $ y = \frac{b}{a}x $, то для вех точек $ M(x, y) $ гиперболы $ \lim_{x \to +\infty} \limits \rho(M, L) = 0$

### Дирректориальное своство эллипса

Число $ \varepsilon = c / a $ называется эксцентриситетом гиперболы. Из определения $ \varepsilon > 1 $, при этом $$ \varepsilon = \sqrt{1 + \frac{b^2}{a^2}}, \space\space\space\space \frac{b}{a} = \sqrt{\varepsilon^2 - 1} $$

И $$ \begin{cases}r_1 = -(a + \varepsilon x) \newline r_2 = -(a - \varepsilon x) \end{cases} $$ $$\begin{cases}r_1 = a + \varepsilon x \newline r_2 = \varepsilon x - a \end{cases}$$

Для левой и праой ветвей соответсвенно.

Прямые $ d_1, d_2 $, заданные в канонической сситеме координат уравнениями

$$ d_1: \space x = -\frac{a}{\varepsilon}, \space\space d_2: \space x = \frac{a}{\varepsilon} $$

называются дирректричами гиперболы

### Теорема 2

Гипербола есть геометрическое место точек $ M $ плоскости, для которых отношение расстояния от данной точки $ F $ к расстоянию до данной прямой $ d $, не проходящей через точку, есть данное полодительное число больше 1, т.е.

$$ \frac{\rho(M, F)}{\rho(M, d)} = \varepsilon, \space\space \varepsilon > 1 $$

### Доказательство

Сводится к рассмотрению расстояний от фокуса до точки и от дирректрисы до точки.
