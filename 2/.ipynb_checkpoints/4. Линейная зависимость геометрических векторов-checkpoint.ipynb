# 4. Линейная зависимость геометрических векторов. Коллинеарные и компланарные вектора 

## Линейная зависисмость

Пусть $ a_1, \dots, a_n $ - вектора линейного пространства $ V $, $ \alpha_1, \dots, \alpha_n \int \mathbb{R} $. Вектор $ \alpha_1 \cdot a_1 + \dots + \alpha_n \cdot a_n $ называется линейной комбинацией векторов $ a_1, \dots, a_n $ c коэффицентами $ \alpha_1, \dots, \alpha_n $. Если вектор $ b $ является линейной комбинацией этих векторов, то говорят, что вектор $ b $ линейно выражается через вектора $ a_1, \dots, a_n $, при этом представление $ b = \alpha_1 \cdot a_1 + \dots + \alpha_n \cdot a_n $ называется линейным разлодением $ b $ по векторам $ a_1, \dots, a_n $. Нулевой вектор выражается через любой вектор и через него не может выражаться ни один ненулевой вектор.  

Линейная комбинация называется тривиальной, если все коэффиценты равны нулю и, соответственно, нетривиальной, если хотя бы один из коэффицентов не равен нулю.  

Если существует нетривиальная линейная комбинация сситемы векторов, равная нулевому вектору, то такая система называется линейно зависимой.

### Теоремы

- Система из одного вектора линейно зависима тогда и только тогда, когда этот вектор - нулевой.
- Система векторов $ a_1, \dots, a_k, \space k > 1 $ линейно зависима тогда и только тогда, когда хотя бы один из её векторов выражается через другие.
- Если подсистема системы векторов линейно зависима, то и вся система линейно зависима.
- Любая подсистема линейно независимой системы лиинейно независима.
- Система векторов $ a_1, \dots, a_k $ линейно независима тогда и только тогда, когда любой вектор, являющийся линейной комбинацией этих векторов, имеет единственное разложение по этим векторам
- Если система векторов $ a_1, \dots, a_k $ линейно независима, а система $ a_1, \dots, a_k, b $ линейно зависима, то вектор $ b $ линейно выражается через вектора $ a_1, \dots, a_k $

### Доказательства

Довольно простые, так что для экономии времени и места не приводятся.

## Геометрический смысл линейной зависимости

Рассмотрим геометрический смысл векторов на прямой, плоскости и в пространстве. Для этого докажем 2 утверждения:

### Утверждение 1

На прямой (плоскости и в пространстве) существует ненулевой вектор (два неколлинеарных и три некомпланарных вектора соответственно).

### Доказательство

В случае прямой достаточно взять две носовпадающие точки $ A, O $, тогда вектор $ a = \overrightarrow{OA} \neq 0 $.  
На плоскости достаточно взять три точки $ O, A, B $, не лежащие на одной прямой тогда вектора $ a = \overrightarrow{OA}, b = \overrightarrow{OB} $ некомпланарны.  
В пространстве достаточно взять четыре точки $ O, A, B, C $, не лежащие в одной плоскости. Тогда вектора $ a = \overrightarrow{OA}, b = \overrightarrow{OB}, c = \overrightarrow{OC} $ не компланарны.

### Утверждение 2

На прямой (плоскости, в пространстве) любой вектор выражается через любой ненулевой вектор (два неколлинеарных вектора и три некомпланарных вектора соответственно).

### Теорема 1

Два вектора линейно зависимы тогда и только тогда, когда они коллинеарны.

### Доказательство. Необходимость  

Пусть они зависимы. Тогда из теорем, перечисленных ранее $ b = \alpha \cdot a \Rightarrow $ они коллинеарны

### Достаточность

Пусть они коллинеарны. Отложим их от одной точки. Тогда согласно утверждению 2, $ b = \alpha \cdot a $

ч.т.д.

### Теорема 2

Три вектора линейно зависимы тогда и только тогда, когда они компланарны.

### Доказательство. Необходимость

Пусть $ c = \alpha \cdot a + \beta \cdot b$. Если вектора $ a, b $ коллинеарны, то вектора $ a, b, c $ коллинеарны, а тем более компланарны. Иначе постром параллерограм из $ a, b $. По правилу параллелограмма $ c $ - его диагональ, принадлежащая плоскости, образованной $ a, b \Rightarrow a, b, c$ - компланарны.

### Достаточность

Допустим вектора $ a, b, c $ компланарны. Отложим их от одной точки. Из утверждения 2 получаем $ c = \alpha \cdot a + \beta \cdot b $  

ч.т.д.

### Теорема 3

Любые 4 вектора линейно зависимы.

### Доказательство

Даны вектора $ a, b, c, d $. Допустим $ a, b, c $ некомпланарны (иначе вся система линейно зависима). Отложим все вектора от одно точки. Тогда из утверждения 2 вытекает, что $ d = \alpha \cdot a + \beta \cdot b + \gamma \cdot c $.  

ч.т.д.
