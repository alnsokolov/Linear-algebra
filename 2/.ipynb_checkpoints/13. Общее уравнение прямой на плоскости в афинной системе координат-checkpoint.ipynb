# 13. Общее уравнение прямой на плоскости в афинной системе координат. Критерий параллельности вектора прямой

### Теорема 1

Линия на плоскости является прямой тогда и только тогда, когда она представляет собой алгебраическую линию первого порядка.

### Доказательство. Необходимость

Пусть $ l $ - прямая на плоскости, проходящая череез точку $ M_0 $ и параллельная ненулевому вектору $ a $. Пусть $ Oxy $ - произвольная афинная система координат и $ a = \{ m, n \}, M_0(x_0, y_0) $. Тогда прямая описывается каноническим уравнением $ \left( \frac{x - x_0}{m} = \frac{y - y_0}{n} \right) $, или, то же самое, уравнением $ \space n(x - x_0) - m(y - y_0) = 0 $, которое, если положить $ A = n, B = -m, C = -nx_0 + my_0 $, может быть записано в виде  

$$ Ax + By + C = 0 $$  

Так как вектор $ a = \{ m, n \} \neq 0 $, то по крайней мере один из коэффицентов $ A $ или $ B $ отличен от нуля. Поэтому левая часть уравнения представляет собой алгебраический многочлен первой степени. Следовательно, любая прямая на плоскости является алгеюраической линией первого порядка.

### Достаточность

Пусть в афинной системе координат $ Oxy $ линия $ l $ определяется уравнением $ Ax + By + C = 0 $. Это уравнение умеет частное решение $ x_0 = -\frac{AC}{A_2 + B_2}, y_0 = -\frac{BC}{A_2 + B_2} $, ибо $ Ax_0 + By_0 + C = 0 $. Вычитая последнее равенство из общего уравнения прямой, получим, что $ A(x - x_0) + B(y - y_0) = 0 $, что то же самое,  
$$ \left | \begin{align} x - x_0 && y - y_0 \newline -B && A \end{align} \right | = 0 $$  

Это уравнение определяет прямую, проходящую через $ M_0{x_0, y_0} $ с направляющим вектором $ a = \{ -B, A \} $

Уравнение $ Ax + By + C = 0 $ называется общим уравнением плоскости. Вектор $ n = \{ A, B \} $ назывется вектором нормали к прямой относительно общего уравнения.

### Теорема 2

В афинной системе координат $ Oxy $ на плоскости вектор $ a = \{ m, n \} $ параллелен прямой, заданной общий уравнением тогда и только тогда, когда $ Am + Bn = 0 $

### Доказательство

Как следует из доказательства теоремы 1 вектор $ b = \{-B, A\} $ является направляющим вектором прямой. Это означает, что вектор $ a $ параллелен этой прямой тогда и только тогда, когда $ a $ коллинеарен $ b $, т.е. $ \left| \begin{align} m && n \newline -B && A \end{align} \right| = 0 $, или, то же самое, $ Am + Bn = 0 $
