# 8. Векторное произведение векторов

Два юазиса пространства $ V $ назувают одинаково ориентированными, если матрица перехода из одного базиса в другой, имеет положительный определитель. Иначе они называются противоположно ориентированными.

### Теорема 1

Отношение одинаковой ориентированности является отношением эквивалентности на множестве всех базисов прострнства $ V $.

### Доказательство

Рефлексивность доказывется в силу того, что переход от одного базиса к самому себе совершается через единичную матрицу, симметричность - мз теоремы об обратной матрица перехода, и того факта, что $ |C^{-1}| \cdot |C| > 0 $, транзитивность - из того, что если $ e' = eC, e'' = e'D $, то $ e'' = e(CD) $, при этом $ |CD| = |C|\cdot|D| > 0 $

Этой операцией множество всех базисов разбивается на два непересекающихся класса. Один из данных классов называют классом правых (полодительно ориентированных), а другой - классом левых (отрицательно ориентированных). Каждый из этих классов называется ориентацией пространства. Вещественное линейное пространство с выбранной на нём ориентацией называется ориентированным пространством. Для того, чтобы задать ориентацию пространства, можно выбрать какой-либо базис и объявить все одинаково ориентированные с ним классы положительно ориентированными.

Обычно класс правых базисов на плоскости $ V_2 $ и в пространстве $ V_3 $ выбирают следующим образом:
- Упорядоченную пару неколлинеарных векторов $ e_1, e_2 $ плоскости называют правой, если кратчайший поворот от $ e_1 $ к $ e_2 $ выполняется против часовой стрелки. В противном случае - левой.
- Упорядоченную тройку некомпланарных векторов $ e_1, e_2, e_3 $ пространства называют правой, если из конца вектора $ e_3 $ кратчайший поворот от $ e_1 $ к $ e_2 $ виден совершающимся против часовой стрелки, и левой - в противном случае.

## Векторное произведение

Пусть в пространстве $ V_3 $ выбрана ориентация. Базисы, задающие эту ориентацию, назовём правыми.  

Векторным произведением вектором $ a $ и $ b $ называется такой вектор $ c $, что:
1. $ |c| = |a| \cdot |b| sin \space (\widehat{a, b}) $
2. $ c $ ортогонален к каждому из векторов $ a $ и $ b $  
И, если $ c \neq 0 $, то
3. $ c $ направлен так, что тройка векторов $ a, b, c $ - правая

Обозначение: $ [a, b] $

### Теорема 2 (критерий коллинеарности)

Векторы $ a $ и $ b $ коллинеарны тогда и только тогда, когда $ [a, b] = 0 $

### Доказательство

Через угол или длину, равные нулю.

### Замечания

1. $ |[a, b]| = S_{ab} $, где $ S_{ab} $ - площадь параллелограмма, построенного на векторах $ a $ и $ b $
2. вектор $ [a, b] \bot \pi(a, b) $ - плоскость, заданная $ a $ и $ b $

### Теорема 3

Векторное произведение антикоммутатично. $ [a, b] = -[b, a] \space \space \forall a, b $

### Доказательство

Через то, что стройка векторов $ a, b, [b, a] $ - левая
