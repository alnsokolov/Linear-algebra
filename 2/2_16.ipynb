# 16. Пусчок прямых на плоскости и плоскостей в пространстве

Множество всех прямых плоскости, проходящих через данную точку $ M_0 $ назввается пучком прямых с центром в точке $ M_0 $. Обозначение: $ \pi(M_0) $

Пусть $ l_1, l_2 $ - две несовпадающие прямые пучка $ \pi(M_0) $ заданные уравнениями $ l_i: \space \space A_i x + B_i y + C_i = 0, \space\space\space\space A_i^2 + B_i^2 \neq 0, \space\space\space\space i = 1, 2 $ в некоторой афинной системе координат $ Oxy $. Положим $ F_i(x, y) = A_i x + B_i y + C_i, i = 1, 2 $

### Теорема 1

Прямая принадлежит пучку $ \pi(M_0) $ тогда и только тогда, когда она определяется уравнением $ \alpha F_1(x, y) + \beta F_2(x, y) = 0, \space \space \alpha, \beta \in \mathbb{R} $

### Доказательство. Достаточность

Очевидна, так как это уравнение определяет линию первого порядка, проходящую через $ M_0(x_0, y_0) $, т.к. $ F_1(x_0, y_0) = F_2(x_0, y_0) = 0 $

### Необходимость

Пусть прямая $ l \in \pi(M_0) $. Возьмём на ней точку $ M_1(x_1, y_1) $, отличную от $ M_0 $. Положим $ \alpha = A_2 x_1 + B_2 y_1 + C_2, \space \beta = -(A_1 x_1 + B_1 y_1 + C_1) $. Так как $ M_1 $ не совпадает с $ M_0 $, хотя бы одно из чисел $ \alpha, \beta $ не равно 0. Тогда уравнение $ \alpha F_1(x, y) + \beta F_2(x, y) = 0 $ определяет прямую, проходящую через $ M_0 $ и $ M_1 $. Следовательно, эта прямая совпадает с $ l $.

Итак, любая прямая пучка $ \pi(M_0) $ определяется двумя пересекающимися прямыми $ l_1, l_2 $ этого пучка. Каждая пара чисел $ \alpha, \beta $, где $ \alpha^2 + \beta^2 \neq 0 $ определяет единственную прямую пучка. Уравнение $ \alpha F_1(x, y) + \beta F_2(x, y) = 0 $ называется уравнением пучка прямых, проходящих через точку пересечения прямых.

Множество всех плоскостей, проходящих через прямую $ l $ называется пучком плоскостей с осью $ l $. Обозначение: $ \pi(l) $  

Пусть $ \pi_1, \pi_2 $ - две пересекающиеся плоскости пучка $ \pi(l) $, заданные уравнениеями $ \pi_i: \space \space A_i x + B_i y + C_i z + D_i = 0, \space\space\space\space i = 1, 2 $ в некоторой афинной сситеме координат $ Oxyz $. Положим $ F_i(x, y, z) = A_i x + B_i y + C_i z + D_i, \space\space i = 1,2 $

### Теорема 2

Плоскость принадлежит пучку $ \pi(l) $ тогда и только тогда, когда она определяется уравнением $ \alpha F_1(x, y, z) + \beta F_2(x, y, z) = 0, \space\space \alpha, \beta \in \mathbb{R}$

### Доказательство

Аналогично случаю с прямыми, $ M_1 $ - люьая точка, не лежащая на оси $ l $.

Данное уравнение называется уравнением пучка плоскостей, проходящих через прямую пересечения плоскостей.
