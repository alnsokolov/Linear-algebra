# 5. Афинная система координат на прямой, плоскости и в пространстве. Преобразование координат

Пусть в пространстве $ V_3 $ (плоскости $ V_2 $ или прямой $ V_1 $) зафиксирована точка $ O $, назваемая полюсом. Для любой точки $ A $ вектор $ r_A = \overrightarrow{OA} $ называется радиус-вектором точки относительно полюса. Задание её радиус-вектором определяет однозначное биективное отображение.  

Если в пространстве зафиксирована точка $ O $ и базис $ e_1, e_2, e_3 $, то говорят, что в пространстве задана афинная система координат (общая декартова) $ \{ ); e_1, e_2, e_3 \} $. Точка $ O $ назвается началом координат; оси, заданные базисом называются осями координат и обозначаются $ Ox, Oy, Oz $ (оси ординат, абсцисс и аппликат соответственно). Плоскость, образуемая осями $ Ox, Oy $ называается координатной плоскостью и обозначается $ Oxy $. Афинная система координат обозначается $ Oxyz $.  

Координатами точки $ A $ называется радиус-вектор $ r $ в базисе $ e_1, e_2, e_3 $. Тот факт, что точка имеет координаты $ x, y, z $ означает, что $ r  = xe_1 + ye_2 + ze_3 $. Обозначается это $ A(x, y, z) $

### Теорема 1

Если точки $ A(x_A, y_A, z_A), B(x_B, y_B, z_B) $ - точки пространства, заданные своими координатами в системе координат $ \{ O; e_1, e_2, e_3 \} $, то вектор $ r = \overrightarrow{AB} = {x_B - x_A, y_B - x_A, z_B - x_A} $

### Доказательство

Через радиус-вектора

### Деление отрезка в данном отношении

Если точка $ M $ делит отрезок $ AB $ в отношении $ \lambda $, то пишут $ (ABM) = \lambda $

### Теорема 2

Пусть $ A(r_1), B(r_2), M(r_3) $ - точки пространства и $ (ABM) = \lambda $. Тогда $$ r_3 = \frac{r_1 + \lambda r_2}{1 + \lambda} $$

## Преобразование координат

Пусть в пространстве даны две афинные системы координат: $ {O; e_1, e_2, e_3}, {O'; e_1', e_2', e_3'} $. Первую назовём старой, а вторую - новой. Также нам известно положение новой системы координат относительно старой: $ O'(\alpha, \beta, \gamma) $ и $ C = \{ c_{ij} \} $ - матрица перехода от базиса $ e $ к базису $ e' $.

Пусть $ M $ - произвольная точка пространства. Обозначим через $ (x, y, z) $ её координаты в старой системе координат, а через $ (x', y', z') $ - в новой, через $ r, r', r_O $ - радиус векторы в старой и новой системе координат и положение $ O' $ относительно $ O $ соответственно. Тогда $ r = r_O + r' $ и  

$ \begin{bmatrix} x \newline y \newline z \end{bmatrix} = \begin{bmatrix} \alpha \newline \beta \newline \gamma \end{bmatrix} + C \begin{bmatrix}x' \newline y' \newline z' \end{bmatrix} \space \space \space $ или $ \space \space \begin{align} x = \alpha + c_{11}x' + c_{12}y' + c_{13}z' \newline y = \beta + c_{21}x' + c_{22}y' + c_{23}z' \newline x = \gamma + c_{31}x' + c_{32}y' + c_{33}z' \end{align} $

Эти соотношения называется формулами преобразования координат. Они выражает старые координаты через новые.
