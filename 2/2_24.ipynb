# 24. Эллипс. Фокусы и дирректрисы

Эллипсом называется геометрическое мето точек $ M $, для которых сумма расстояний до двух фиксированных точек $ F_1, F_2 $ плоскости есть постоянное число, большее, чем расстояние мажду $ F_1 $ и $ F_2 $. Это число обозначим за $ 2a $. Точки $ F_1, F_2 $ называются фокусами эллипса, расстояние между ними называется фокусным расстоянием и обозначается через $ 2c $. Числа $ r_1 = \rho(M, F_1), r_2 = \rho(M, F_2) $ называются фокальными радиусами точки $ M $.  

Из определения следуеч, что точка плоскости является точкой эллипса тогда и только тога, когда $ r_1 + r_2 = 2a, a > c $

Введём на плоскости прямоугольную декартову систему координат $ Oxy $, приняв за начало $ O $ середину отрезка $ [F_1 F_2] $, за ось $ Ox $ - прямую $ F_1 F_2 $, ориентированную от $ F_1 $ к $ F_2 $. Ориентацию на оси $ Oy $ выбираем произвольно. Эта сситема координат называется канонической сситемой координат для данного эллипса.  

Фокусы имеют в ней координат $ F_1(-c, 0), F_2(c, 0) $.

### Теорема 1

Уравнение эллипса в канонической сситеме координат имеет вид $$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1, \space\space\space\space a \geqslant b > 0 $$

### Доказательство

$ \sqrt{(x + c)^2 + y^2} + \sqrt{(x - c)^2 + y^2} = 2a $. Пустём преобразований это уравнение приводится к придедённому в формулироваке теоремы

Приведённое уравнение называется кононическим ураынением эллипса. В случае $ a = b $ эллипс превращается в окружжность радиуса $ a $ с центром в точке $ O = F_1 = F_2 $

### Свойства эллипса:

1. Координатные оси являются осями симметриии эллипса, начало координат - центром симметрии эллепса. Числа $ 2a, 2b $ называются большой и малой осями эллипса, а $ a, b $ - большой и малой полуосями эллипса.
2. Все точки эллипса лежат в прямоугольнике, ограниченном прямыми $ x = \pm a, y = \pm b $. Тогки пересечения эллипа с этими прямыми нызываются вершинами эллипса
3. Исследуем форму эллипса в первой четверти координатной системы. Пусть $ y(x) = \frac{b}{a}\sqrt{a^2 - x^2} $
   - $ y(x) $ непрерывна и убывает от $ y = b $ до $ y = 0 $
   - $ y'(0) = 0, y'(a) = \infty \Rightarrow $ касательная к эллипсу в точке $ x = 0 $ параллельна оси абсцисс, а в точке $ x = a $ - оси ординат
   - $ y''(x) < 0 \Rightarrow $ линия выпукла вверх

### Дирректориальное своство эллипса

Число $ \varepsilon = c / a $ называется эксцентриситетом эллипса. Из определения $ 0 \leqslant \varepsilon < 1 $, при этом $$ \varepsilon = \sqrt{1 - \frac{b^2}{a^2}}, \space\space\space\space \frac{b}{a} = \sqrt{1 - \varepsilon^2} $$

И $ r_1 = a + \varepsilon x, r_2 = a - \varepsilon x$

Чем меньше эксцентриситет эллипса, тем ближе он к окружности.

Для эллипса, не являющегося окружностью, вре прямые $ d_1, d_2 $, заданные в канонической сситеме координат уравнениями

$$ d_1: \space x = -\frac{a}{\varepsilon}, \space\space d_2: \space x = \frac{a}{\varepsilon} $$

Называются дирректричами эллипса

### Теорема 2

Эллипс, не являющийся окружностью, есть геометрическое место точек $ M $ плоскости, для которых отношение расстояния от данной точки $ F $ к расстоянию до данной прямой $ d $, не проходящей через точку, есть данное полодительное число меньше 1, т.е.

$$ \frac{\rho(M, F)}{\rho(M, d)} = \varepsilon, \space\space 0 < \varepsilon < 1 $$

### Доказательство

Сводится к рассмотрению расстояний от фокуса до точки и от дирректрисы до точки.
