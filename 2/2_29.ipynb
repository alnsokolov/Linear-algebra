# 29. Эллипсоиды, гиперболоиды, параболоиды, косинусы и цилиндрические поверхности

### Теорема 1

Для любой алгебраической поверхности второго порядка существует прямоугольная декартова сситема координат $ Oxyz $, в которой уравнение этой поверхности имеет один и лишь один из следующих видов:

1. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 1 $ - эллипсоид
2. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = -1 $ - мнимый эллипсоид
3. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} + \frac{z^2}{c^2} = 0 $ - вырожденный эллипсоид
4. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1 $ - однополюсный гиперболоид
5. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = -1 $ - двуполюсный гиперболоид
6. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 0 $ - конус
7. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 2z $ - элиптический параболоид
8. $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 2z $ - гиперболический параболоид
9. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $ - элиптический цилиндр
10. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = -1 $ - мнимый элиптический цилиндр
11. $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $ - гиперболический цилиндр
12. $ y^2 = 2px $ - параболический цилиндр
13. $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 0 $ - пара пересекающихся плоскостей
14. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 0 $ - пара мнимых пересекающихся плоскостей
15. $ y^2 = a^2 $ - пара параллельных плоскостей
16. $ y^2 = -a^2 $ - пара мнимых параллельных плоскостей
17. $ y^2 $ - пара совпадающих плоскостей

### Доказательство

Перебираем все возможные $ \lambda_i, i = \overline{1,3} $ в приведенных уравнениях линии второго порядка, видим, что больше вариантов возможных линий второго порядка нет.

## Геометрические свойства

#### 1.Эллипсоид
Числа $ a, b, c $ в каноническом уравнении называются полуосями эллипсоида. Координатные плоскости являются плоскостями симметрии, а начало координат - центром симметрии. Эллипсоид ограничен параллелепипедом $ |x| \leqslant a, |y| \leqslant b, |z| \leqslant c $

#### 2. Мнимый эллипсоид
Точек, удовлетворяющих этому уравнению, в пространстве нет. Это уравнение принято называть мнимым эллипсоидом.  

(далеше для краткости буду писать просто "точек нет" под названием)

#### 3. Вырожденный эллипсоид
Уравнению удовлетворяет только начало координат.

#### 4. Однополюсный гиперболоид
Координатные плоскости являются плоскостями симметрии, а начало координат - центром симметрии. Сечения гиперболоида плоскостями $ z = h $ представляют собой эллипсы с полуосями $ a(1 + h^2 / c^2)^{1/2} $ и $ b(1 + h^2 / c^2)^{1/2} $, неограниченно возрастающие при $ h \to \infty $. Эллипс, получающийся при $ h = 0 $ называют горловым эллипсом гиперболоида.

Плоскость $ y = h $ пересекает гиперболоид:
1. при $ |h| < b $ по гиперболе с полуосями $ a(1 - h^2 / c^2)^{1/2}, c(1 + h^2 / c^2)^{1/2} $, которые убывают от $ a $ и $ c $ до нуля при $ |h| \to b $
2. при $ |h| = b $ по двум пересекающимся прямым
3. при $ |h| > b $ по гиперболе с полуосями $ c(-1 + h^2 / c^2)^{1/2}, a(-1 + h^2 / c^2)^{1/2} $, которые неограниченно возрастают при $ |h| \to \infty $

Сечения гиперболоида плоскостями $ x = h $ обладают теми же свойствами.  
Важной особенностью однополюсного гиперболоида является наличие прямых, полность лежащих на этой поверхности (это следующий билет)

#### 5. Двуполюсный гиперболоид
Координатные плоскости являются плоскостями симметрии, а начало координат - центром симметрии.

Плоскости $ z = h $ пересекают этот гиперболоид:
1. при $ |h| > c $ по эллипсам, размеры которых неограниченно возрастают при $ |h| \to \infty $
2. при $ |h| = c $ в единственной точке

В слое между плоскостями $ h = c, h = -c $ нет ни одной точки гиперболоида. Таким образом, двуполюсной гиперболоид состоит из двух симметричных полостей, расположенных в полупространствах $ z \geqslant c $ и $ z \leqslant -c $

Плоскости $ x = h $, $ y = h $ пересекают его по гиперболам

#### 6. Конус

Координатные плоскости являются плоскостями симметрии, а начало координат - центром симметрии.

Плоскости $ z = h, h \neq 0 $ пересекают конус по эллипсам, размеры которых неограниченно возрастают при $ |h| \to \infty $.
Плоскости $ x = h, y = h, h \neq 0 $ пересекают конус по гиперболам с полуосями $ c|h| / b, a|h| / b $ и $ c|h| / a, b|h| / a $ соответственно. Плоскости $ x = 0, y = 0 $ - по парам пересекающихся прямых.
Плоскими сечениями конуса являются и параболы. Параболой будет сечение конуса плоскостью $ z = h + cx / a, h \neq 0 $, ибо числа $ x, y $ являются на этой плоскости афинными координатами, а уравнение линии, высекаемой на ней конусом, имеет в этих координатах вид
$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{(h + cx/a)^2}{c^2} = 0 \Leftrightarrow y^2 = 2\frac{hb^2}{ac}(x + \frac{ha}{2c}) $$
Таким образом и эллипс, и гипербола, и парабола являются плоскими сечениями конуса. На этом основании эти линии обычно называют коническими сечениями.

#### 7. Элиптический параболоид
Плоскости $ x = 0, y = 0 $ являются плоскостями симметрии. Центра симметрии у него нет. Плоскости $ z = h, h > 0 $ пересекают параболоид по эллипсам, размер которых неограниченно возрастает при $ h \to \infty $. Плоскости $ x = h, y = h $ пересекают параболоид по параболам, ветви которых направлены вверх

#### 8. Гиперболический параболоид

Плоскости $ x = 0, y = 0 $ являются плоскостями симметрии. Центра симметрии у него нет. Плоскости $ y = h, x = h $ пересекают параболоид по параболам, ветви первой параболы направлены вверх, а второй - вниз.

Плоскости $ z = h $ пересекают параболоид:
1. при $ h < 0 $ по гиперболам $ -\frac{x^2}{a_1^2} + \frac{y^2}{b_1^2} $, где $ a_1^2 = -2a^2h, b_1^2 = -2b^2h $
2. при $ h > 0 $ по гиперболам $ \frac{x^2}{a_2^2} - \frac{y^2}{b_2^2} $, где $ a_2^2 = 2a^2h, b_2^2 = 2b^2h $
3. при $ h = 0 $ по паре пересекающихся прямых
