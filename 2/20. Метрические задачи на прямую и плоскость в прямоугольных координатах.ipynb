# 20. Метрические задачи на прямую и плоскость в прямоугольных координатах

Пусть $ Oxyz $ - прямоугольная декартова система координат в пространстве.

1. Углом мжду двумя пряммыми в пространстве называется любой из углов между параллельными им прямыми, проходящими через какую-либо точку пространства. Таким образом, две прямые в пространстве образуют два различных угла, сумма которых равна $ \pi $. Очевидно, что угол между направляющими векторами равен одному из этих углов. Следовательно, угол $ \varphi $ между прямыми $ l_i: r = r_i + ta_i, i = 1, 2 $ совпадает с углом между их направляющими векторами $ a_i = \{ m_i, n_i, k_i \} $, вычисляется согласно формуле  
$$ cos \space \varphi = \frac{m_1 m_2 + n_1 n_2 + k_1 k_2}{\sqrt{m_1^2 + n_1^2 + k_1^2}\sqrt{m_2^2 + n_2^2 + k_2^2}} $$  

2. Угол между прямой и плоскостью называется менбший из угловмежду прямой и её ортогональной проекцией на эту плоскость. Если прямая и плоскость перпендикулярны, то угол между ними считается равным $ \frac{\pi}{2} $. Угол $ \varphi $ между прямой $ l:r + r_0 + ta $ и плоскостью $ \pi: Ax + By + Cz + D = 0 $ находится как дополнительный к углу между направляющим вектором прямой и вектором нормали к плоскостии вычисляется согласно формуле  
$$ sin \space \varphi = \frac{|Am + Bn + Ck|}{\sqrt{m^2 + n^2 + k^2}\sqrt{A^2 + B^2 + C^2}}, \space\space\space\space 0 \leqslant \varphi \leqslant \frac{\pi}{2} $$

3. Расстояние $ \rho(M_1, l) $ от точки $ M_1(r_1) $ до прямой $ l: r = r_0 + ta $ находится как высота $ h $ параллелограмма, построенного на векторах $ a $ и $ \overrightarrow{M_0M_1} $, площадь и основание которого известны:  

$$ \rho(M_1, l) = \frac{|[a, r_1 - r_0]|}{|a|} $$

4. Расстояние между скрещивающимися прямыми $ l_i: r = r_i + ta_i, i = 1, 2 $ называется расстояние между параллельными плоскостями, в которых лежат прямые $ l_1 $ и $ l_2 $. Это расстояние $ \rho(l_1, l_2) $ находится как высота параллелепипеда, построенного на векторах $ \overrightarrow{M_1M_2}, a_1, a_2 $, объём и площадь основания которого известны:  

$$ \rho(l_1, l_2) = \frac{|(r_2 - r_1, a_1, a_2)|}{|[a_1, a_2]|} $$
