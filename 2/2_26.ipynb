# 26. Парабола. Фокус и дирректриса

Параболой называется геометрическое место точек плоскости, лоя которых расстояние от некоторой фиксированной точки $ F $ равно расстоянию до некоторой фиксированной прямой $ d $, не проходящей через точку. Точка $ F $ называется фокусом параболы, а прямая $ D $ - её дирректрисой. Расстояние от фокуса параболы для её дирректрисы называется фокальным расстоянием параболы и обозначается за $ p $. Эксцентриситет параболы по определению считается равным 1.

### Теорема 1

Уравнение для параболы в каноничсекой системе координат $ Oxy $ имеет вид $$ y^2 = 2px, p > 0 $$

### Доказательство

Выполняется ассмотрением расстояния от точки до фокуса и до дирректрисы.

### Свойства параболы:

1. Ось $ Ox $ канонической системы координат является осью симметрии параболы. Она называется осью параболы. Начало координат лежит на параболе и называется вершиной параболы.
2. Все точки параболы лежат в правой полуплоскости от оси $ Oy $
3. Исследуем форму параболы в первой четверти координатной системы. Пусть $ y(x) = \sqrt{2px}, x \geqslant 0 $
   - $ y(x) $ непрерывна и возрастает на $ [0, +\infty) $, причём $ y(0) = 0, \lim_{x \to +\infty} \limits y(x) = +\infty $
   - $ y'(0) = \infty  \Rightarrow $ касательная к параболе в точке $ x = 0 $ параллельна оси ординат
   - $ y''(x) < 0 \Rightarrow $ линия выпукла вверх

### Замечание

Определение параболы, по существу, означает её дирректориальное свойство  

$$ \frac{\rho(M, F)}{\rho(M, d)} = \varepsilon $$

Где $ F $ - её фокус, $ d $ - её дирректриса, $ \varepsilon $ - эксцентриситет
