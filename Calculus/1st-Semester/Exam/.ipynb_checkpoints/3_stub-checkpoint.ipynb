### Определение 3

### Утверждение 1

Любое подмножество счётного множества пусто, конечно или счётно (такие множества называются не более чем счётными).

### Доказательство

Пусть множество $ A $ счётно, $ B \subset A $. Занумеруем элементы $ A $, после чего пронумеруем элементы множества $ B $ в порядке возрастания номеров. Если процесс прервётся, то $ B $ - конечно, иначе $ B $ - счётно.
