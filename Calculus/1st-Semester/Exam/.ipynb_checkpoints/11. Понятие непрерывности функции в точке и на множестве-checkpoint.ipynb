# 11. Понятие непрерывности функции в точке и на множестве. Арифметические операции над непрерывными функциями. Классификация точек разрыва.

Пусть $ f $ определена на множестве $ X, a $ - пердельная точка множества $ X $.

### Определение 1 (формальное)

Пусть $ a \in X $. Функция $ f $ непрерывна в $ a $, если $ \lim_{x \to a}\limits f(x) = f(x) $

### Определение 2 (Коши)

Функция $ f $ непрерывна в точке $ a $, если $ \forall \varepsilon > 0 \space \exists \delta(\varepsilon) > 0: \space \forall x \in \dot B_\delta(a) \cap X \space |f(x) - f(a)| < \varepsilon $

### Определение 3 (Гейне)

Функция $ f $ непрерывна в точке $ a $, если для любой последовательности $ \{ x_n \} $, т.ч. $ \lim_{n \to \infty}\limits x_n = a $ выполнено $ \lim_{n \to \infty}\limits f(x_n) = f(x) $

Определения 1-3 эквивалентны.

### Определение 4

Пусть $ f(x) $ определена на множестве $ X $, $ a $ - предельная точка $ X $. ФУнкция $ f $ разрывна в точке $ a $, если она не является непрерывной в данной точке.

### Определение 5

Пусть $ f(x) $ определена на множестве $ X $. Точка $ a $ является предельной точкой множеств $ (a - \delta, a) \cap X $ и $ (a, a + \delta) \cap X, \delta > 0 $. Функция непрерывна в точке $ a $ справа (слева), если $ \lim_{x \to a + 0}\limits f(x) = f(a) \space $ ($\lim_{x \to a - 0}\limits f(x) = f(a)$ )

### Утверждение 1

Пусть $ f(x) $ определена в $ B_\delta(a), \delta > 0 $. Из того что $ f $ непрерывна в этой точке, следует, что она непрерывна в этой точке слева и справа.

### Доказательство

$ \exists \lim_{x \to a}\limits f(x) = f(a) \Rightarrow \begin{cases} \lim_{x \to a+0}\limits f(x) = f(a) \newline \lim_{x \to a-0}\limits f(x) = f(a) \end{cases} $

## Классификация точек разрыва

1. Устранимые разрывы
   Точка $ a $ называется точкой устранимого разрыва функции $ f $, если $ \exists \lim_{x \to a}\limits f(x) \neq f(a) $
2. Разрыв 1-го рода
   Точка $ a $ называется точной разрыва 1-го рода, если существуют левый и правый предел в точке $ a $, но они не равны.
3. Разрыв 2-го рода
   Точка $ a $ называется точкой разрыва 2-го рода, если ходя бы один из левого и правого предела не существуют, или же один из них равен бесконечности.

## Арифметические операции

### Теорема 1 (об арифметических операциях)

Пусть $ f, g $ определены на множестве $ X $ и непрерывны в точке $ a \in X $. Тогда $ f(x) \pm g(x), f(x)g(x), \frac{f(x)}{g(x)} $ (если $ g(x) \neq 0 $) являются непрерывными в точке $ a $.

### Доказательство

Сразу следует из теоремы об арифметических операциях на функциях, имеющих предел в точке.
