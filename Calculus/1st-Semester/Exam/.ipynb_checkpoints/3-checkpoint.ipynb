# 3. Счетные множества и множества мощности континуум. Неэквивалентность множества мощности континуум счетному множеству.

### Определение 1

Множества, эквивалентные множеству $ \mathbb{N} $ называются счётными и имебт мощность $ \aleph_0 $ (илеф-нуль).

### Опредление 2

Множества, эквивалентные множеству точек на отрезке $ [0, 1] $ имеют мощность континуум или $ c $.

### Аксиома непрерывности

 Для любых множеств действительных чисел $ B $ и $ A $ (все элементы которого не превышают элементов $ B $) существует действительное число $ \xi $, принадлежащее числовому отрезку между наибольшим элементом множества $ A $ и наименьшим элементом множества $ B $.

### Теорема Кантора

Множество мощнсти континуум не является счётным.

### Доказательство

Допусти множество всех точек отрезка $ [0, 1] $ счётно. Значит их можно расположить в последовательности $ x_1, x_2, \dots $.  

Разобьём отрезок $ [0, 1] $ на три равные части. Точка $ x_1 $ не может принадлежать сразу всем отрезкам сразу. Допустим она не принадлежит отрезку $ D_1 $. ВОзьмём его и разделим на три равные части. Среди получившихся отрезков есть отрезок $ D_2 $, не содержащий точку $ X_2 $. Разделим его на три части и т.д. В итоге мы получим последовательность отрезков $ D_1, D_2, \dots $, которые в силу Аксиомы кантора содержат одну общую точку $ x $. Так как она принадлежит всем отрезкам, она не совпадает ни с одной из точек $ x_1, x_2, \dots $, то есть эти точки не исчерпывают всех точек отрезка $ [0, 1] $. Противоречие. Значит множество точек отрезка $ [0, 1] $ не счётно.
