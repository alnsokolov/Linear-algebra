### Задача 1

#### Теорема о пределе монотонной ограниченной последовательности

Если $ \{x_n\} $ является неубывающей и ограниченной сверху (невозрастающей и ограниченной снизу), то она сходится ($ \exists \space \lim\limits_{n \to \infty} x_n $)

$ \triangle \space $ Пусть $ \space x_{n + 1} \geqslant x_n \forall \space n \in \mathbb{N}, \exists M \in \mathbb{R}, $ т.ч. $ x_n \leqslant M \space \forall \space n \in \mathbb{N} \Rightarrow \exists \space \bar{x} = sup \{x_n\}. $ Покажем, что $ \bar{x} = \lim\limits_{n \to \infty} x_n $  

Возьмём $ \varepsilon > 0 $ По определению $ sup \space \exists N \in \mathbb{N}, $ т.ч. $ \bar{x} - \varepsilon < x_n \leqslant \bar{x}. $ В силу монотонности $ \{x_n\}, \forall n \geqslant N: \bar{x} - \varepsilon < x_n \leqslant \bar{x} < \bar{x} + \varepsilon \Rightarrow |x_n - \bar{x}| < \varepsilon \ \forall \space n \geqslant N \Leftrightarrow \lim\limits_{n \to \infty} x_n= \bar{x} $

### Задача 2

#### Определение, свойства и график функции $ y=cos x $

Тестовый LaTeX

$ X = \mathbb{R} $, $ Y = [-1, 1] $  

Функция чётная, периодическая ($ T = 2\pi $).  

Возрастает на $ [\pi + 2 \pi k, 2\pi + 2 \pi k] $, убывает на $ [2 \pi k, \pi + 2 \pi k], k \in \mathbb{Z} $  

Функция непрерывна на $ \mathbb{R} $

```python
x = np.linspace(- 2 * np.pi, 2 * np.pi, 200)
y = np.cos(x)

matplotlib.rcParams.update({'font.size': 18})

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].plot(x, y)
axs[0].set_title('$ y=cos x $')

axs[1].axis('off')
axs[1].text(0, 0.02, r"""
$ X = \mathbb{R} $, $ Y = [-1, 1] $  

Функция чётная, периодическая ($ T = 2\pi $).  

Возрастает на $ [-\frac{\pi}{2} + 2 \pi k, \frac{\pi}{2} + 2 \pi k] $, убывает на $ [\frac{\pi}{2} + 2 \pi k, \frac{3 \pi}{2} + 2 \pi k], k \in \mathbb{Z} $  

Функция непрерывна на $ \mathbb{R} $
""")
plt.show()
```
