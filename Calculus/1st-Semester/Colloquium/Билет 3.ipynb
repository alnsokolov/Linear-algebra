### Задача 1

#### Понятие числовой последовательности. Определения ограниченной и неограниченной, бесконечно большой и бесконечно малой последовательности. Примеры.

Числовая последовательность $ \{x_n\} $ - множество чисел $ \{x_1, x_2, \dots, x_n, \dots \} $, в котором наждому натуральному числу $ n $ поставленов соответствие какое-то вещественное число $ x_n $

Ограниченная последовательность - последовательность, для которой $ \exists A: \forall x \in \{x_n\} \ |x_n| \leqslant A $  
Неограниенная последовательность - оследовательность, для которой $ \forall A: \exists x \in \{x_n\} \ |x_n| > A $

Последовательность называется бесконечно большой, если $ \forall A \ \exists N(A): \forall n \geqslant N \  |x_n| > A $  
Последовательность называется бесконечно малой, если $ \forall \varepsilon \ \exists N(\varepsilon): \forall n \geqslant N \  |x_n| < \varepsilon $

Примеры:  

$ x_n = n $ - бесконечно большая, неограниченная последовательность  
$ x_n = \frac{1}{n} $ - бесконечно малая, ограниченная последовательность

### Задача 2

#### Определение, свойства и график функции $ y=tg x $

Тестовый LaTeX

$ X = (-\frac{\pi}{2} + 2\pi k, \frac{\pi}{2} + 2\pi k) \space k \in \mathbb{Z} $, $ Y = \mathbb{R} $  

Функция нечётная, периодическая ($ T = \pi $).  

Возрастает на $ \mathbb{X} $

Функция непрерывна на $ \mathbb{X} $

```python
x = np.linspace(- 2 * np.pi, 2 * np.pi, 200)
y = np.tan(x)
threshold = 10
y[y>threshold] = np.inf
y[y<-threshold] = np.inf

matplotlib.rcParams.update({'font.size': 18})

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].set_ylim([-7, 7])
axs[0].plot(x, y)
axs[0].set_title('$ y=tg x $')

axs[1].axis('off')
axs[1].text(0, 0.02, r"""
$ X = (-\frac{\pi}{2} + 2\pi k, \frac{\pi}{2} + 2\pi k) \ k \in \mathbb{Z} $, $ Y = \mathbb{R} $  

Функция нечётная, периодическая ($ T = \pi $).  

Возрастает на $ X $

Функция непрерывна на $ X $
""")
plt.show()
```
