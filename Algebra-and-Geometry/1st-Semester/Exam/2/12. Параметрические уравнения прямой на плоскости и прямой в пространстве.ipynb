# 12. Параметрические уравнения прямой на плоскости и прямой в пространстве

## Прямая

Ненулевой вектор, коллинеарный прямой называется её направляющим вектором. Из аксиом геометрии следует, что через любую точку проходит единственная прямая м заданным направляющим вектором.

Пусть прямая $ l $ с направляющим вектором $ a $ проходит через точку $ M_0 $. Очевидно, что точка $ M $ ледит на прямой тогда и только тогда, когда векторы $ MM_0 $ и $ a $ коллинеарны. С учётом того, что $ a \neq 0 $, вектор $ \overrightarrow{M_0M} $ линейно выражается через вектор $ a $: $$ \overrightarrow{M_0M} = ta, t \in \mathbb{R} $$

Пусть $ r = \overrightarrow{OM}, r_0 = \overrightarrow{OM_0} $ - радиус-векторы точек $ M $ и $ M_0 $ относительно полюса $ O $. Тогда $ \overrightarrow{M_0M} = r - r_0 $ и условие линейного выражения данного вектора через $ a $ может быть записано иначе:  
$$ r = r_0 + at, \space t \in \mathbb{R} $$  
Или в координатной форме $ Oxy $:  
$$ \begin{cases} x &= &x_0 + tm \newline y &= &y_0 + tn \end{cases}, \space \space t \in \mathbb{R} $$

Данные уравнения называются параметрическими уравнениями прямой на плоскости в векторной и координатной форме соответственно.

Замечание: число $ t $ является координатой точки $ M $ на прямой $ l $ в система координат $ \{M_0; a\} $

## Плоскость

Пусть плоскость $ \pi $ с направляющими вектоорами $ p_1, p_2 $ проходит через точку $ M_0 $. Очевидно, что точка $ M $ лежит в плоскости $ \pi $ тогда и только тогда, когда векторы $ \overrightarrow{M_0M}, p_1, p_2 $ компланарны, т.е. линейно зависимы. С учётом неколлинеарности векторов $ p_1, p_2 $ это равносильно тому, что вектор $ \overrightarrow{M_0M} $ линейно выражается через вектора $ p_1 $ и $ p_2 $:  
$$ \overrightarrow{M_0M} = up_1 + vp_2, \space \space \space u, v \in \mathbb{R} $$

Аналогично прямой, данное уравнение может быть переписано в виде  
$$ r = r_0 + up_1 + vp_2 $$  
Или в координатной форме:  
$$ \begin{cases} x &= &x_0 + um_1 + vm_2 \newline y &= &y_0 + un_1 + vn_2 \newline z &= &z_0 + uk_1 + vk_2 \end{cases}, \space \space \space u, v \in \mathbb{R} $$  

Данные уравнения называются параметрическим уравнением плоскости в веторной и координатной форме соответственно.  
Замечание: числа $ u, v $ являются координатами точки $ M $ плоскости $ \pi $ в системе координат $ \{ M_0; p_1, p_2 \} $.
