# 17. Полуплоскости и полупространства

Каждая прямая $ l $ на плоскости делит эту плоскость на две полуплоскости, при этом точки $ M_1 $ и $ M_2 $, не лежащие на $ l $ принадлежат одной полуплоскости, если отрезок $ [M_1M_2] $ не имеет отбщих точек с $ l $, в противном случае точки принадлежат разным полуплоскостям.

Пусть $ l $ задаётся уравнением $ Ax + By + C = 0 $

### Теорема 1

Точки $ M_1(x_1, y_1) $ и $ M_2(x_2, y_2) $ принадлежат разным полуплоскостям тогда и только тогда, когда  

$$ (A x_1 + B y_1 + C)(A x_2 + B y_2 + C) < 0 $$  

### Доказательство

Заметим, что $ M_0 $ является внутренней точкой отрезка $ [M_1 M_2] $ тогда и только тогда, когда $ \overrightarrow{M_1M_0} = t \overrightarrow{M_1M_2} $, где $ 0 < t < 1 $, т.е. $ x_0 = (1 - t)x_1 + t x_2, y_0 = (1 - t)y_1 + t y_2, 0 < t < 1 $

Точки $ M_1(x_1, y_1) $ и $ M_2(x_2, y_2) $ принадлежат разным полуплоскостям тогда и только тогда, когда существует точка $ M_0(x_0, y_0) $, общая для прямой $ l $ и отрезка $ [M_1 M_2] $, причём $ M_0 $ является внутренней точкой отрезка $ [M_1 M_2] $, т.е.  
$$ \begin{cases} Ax_0 + By_0 + C = 0 \newline x_0 = (1 - t)x_1 + t x_2, y_0 = (1 - t)y_1 + t y_2, 0 < t < 1 \end{cases} $$

С учётом очевидного тождества $ C = (1 - t)C + tC $, получим, что точки $ M_1 $ и $ M_2 $ принадлежат разным полуплоскостям тогда и только тогда, когда существует число $ t $, такое что  
$$ (1 - t)(A x_1 + B y_1 + C) + t(A x_2 + B y_2 + C) = 0,\space\space 0 < t < 1 $$

Это равносильно тому, что $ (A x_1 + B y_1 + C)(A x_2 + B y_2 + C) < 0 $

Итак для координат $ (x, y) $ одной полуплоскости выполняется неравенство $ A x + B y + C > 0 $, а для другой - $ A x + B y + C < 0 $. Полуплоскость точек, для которой $ A x + B y + C > 0 $ называется положительной полуплоскостью относительно уравнения $ A x + B y + C = 0 $ прямой $ l $ и обозначается символом $ \pi_+ $, а другая полуплоскость называется отрицательной полуплоскостью и обозначается $ \pi_- $

### Теорема 2

Вектор нормали $ n = \{ A, B \} $ к прямой $ l: A x + B y + C = 0 $, отложенный от любой точки плоскости, направлен в сторону положительной полуплоскости.

### Доказательство

Пусть $ M_0(x_0, y_0) $ - произвольная точка прямой $ l $. Отложим вектор $ n $ от точки $ M_0 $, пусть конец вектора совпадёт с точкай $ M_1(x_1, y_1) $. Тогда $ n = \{ x_1 - x_0, y_1 - y_0 \} = \{A, B\} $ и, следовательно, $ x_1 = A + x_0, y_1 = B + y_0 $. Подставив координаты $ x_1, y_1 $ точки $ M_1 $ в левую часть уравнения прямой $ l $, получаем, что $ A x_1 + B y_1 + C = A^2 + B^2 + A x_0 + B x_0 + C = A^2 + B^2 > 0 $, т.к. $ A x_0 + B y_0 + C = 0 $ и $ n \neq 0 $

Аналогичные факты имеют мето и в пространстве. Плоскость $ \pi : Ax + By + Cz + D = 0 $ разбивает пространство на два полупространства

### Теорема 3

Точки $ M_1 (x_1, y_1, z_1) $ и $ M_2 (x_2, y_2, z_2) $ принадлежат разным полупространствам относительно плоскости $ \pi $ тогда и только тогда, когда  
$$ (A x_1 + B y_1 + C z_1 + D)(A x_2 + B y_2 + C z_2 + D) < 0 $$
