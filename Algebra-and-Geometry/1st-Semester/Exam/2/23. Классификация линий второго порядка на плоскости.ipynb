# 23. Классификация линий второго порядка на плоскости

### Теорема 1

Общее уравнение линии второга порядка в прямоугольной декартовой системе координат,определяет одну и только одну из  девяьт линий, для каждой из которых существует прямоугольная ссиетма координат, в которой уравнение линии имеет следующий вид:

1. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 \space \space (a \geqslant b > 0) $ - эллипс
2. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} -1 1 $ = мнимый эллипс
3. $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 0 $ - пара мнимых пересекающихся прямых
4. $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $ - гмпербола
5. $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 0 $ - пара пересекающихся прямых
6. $ y^2 = 2px \space\space (p > 0) $ - парабола
7. $ y^2 = a^2 $ - пара параллельных прямых
8. $ y^2 = -a^2 $ - пара параллельных мнимых прямых
9. $ y^2 = 0 $ - пара совпадающих прямых

### Доказательство

Длиное, муторное $ \dots $

### Метод Лагранжа (метод выделения полных квадратов)

Он заключается в выделении полных квадратов из данного уравнения линии второго порядка. 
1. Пусть в общем уравнении $ a_{11} \neq 0 $. Выделим полный кдарат в группе членов, содержащих переменную $ x $:

$$ a_{11}x^2 + 2a_{21}xy + 2a_{13}x = a_{11}(x + \frac{a_{12}}{a_{11}} + a_{13})^2 - a_{11}(\frac{a_{12}}{a_{11}} + a_{13})^2 $$

Положиим $ x' = x + \frac{a_{12}}{a_{11}} + a_{13} $  
Тогда изначальное уравнение примет вид

$$(*) \space\space\space\space\space\space\space\space a_{11}x'^2 + a_{22}'y^2 + 2a_{23}'y + a_{33} = 0 $$  

Если $ a_{22}' \neq 0 $, то положив $ y' = y + \frac{a_{23}'}{a_{22}'} $, уравнение принимае вид

$$ a_{11}x'^2 + a_{22}'y'^2 + a_{33}'' = 0, \space\space a_{11}a_{22}' \neq 0 $$

которое является приведённым уравнением типа I. Матрица перехода к новому базису выглядит так:

$$ Q = \begin{bmatrix} 1 & -\frac{a_{12}}{a_{11}} \newline 0 & 1 \end{bmatrix} $$

Если в уравнении $ (*) \space\space a_{22}' = 0 $, то:  
а. В случае $ a_{23}' = 0 $ уравнение является приведённым III типа  
б. В случае $ a_{23}' \neq 0 $, положив  
$$ y' = y + \frac{a_{33}'}{2s_{23}'} $$
приводим уравнение $ (*) $ в виду  
$$ a_{11}x'^2 + 2a_{13}' = 0, \space\space\space\space a_{11}a_{13}' \neq 0 $$
которое является приведённым II типа

2. Пусть в общем уравнении $ a_{11} = 0, a_{22} \neq 0 $. Поступая так же, как и в п.1 приходим к приведённым уравнениям
3. Пусть в общем уравнении $ a_{11} = 0, a_{22} = 0 $. Тогда $ a_{12} \neq 0 $. Поло;ив $ x = x' + y', y = x' - y' $ сводим этот случай к ужже рассмотренным.

С помощью этих преобразований уравнение сводится к одному и только одному из 9 уравнений из теоремы 1. С помощью метода Лагранжа можно установить тип линии, но не может быть получено каноническое уравнение.
