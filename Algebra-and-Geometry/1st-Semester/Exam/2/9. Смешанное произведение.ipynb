# 9. Смешанное произведение

Смешанным произведением векторов $ a, b, c $ будем называть скалярное произведение вектороного произведения $ [a, b] $ на $ c $. Обозначение: $ (a, b, c) = ([a, b], c) $

### Теорема 1

Векторы $ a, b, c $ компланарны тогда и только тогда, когда $ (a, b, c) = 0 $.

### Доказательство. Необходимость

Пусть $ a, b, c $ компланарны. Будем считать, что $ a $ и $ b $ не коллинеарны и $ c \neq 0 $ (иначе сразу же $ (a, b, c) = 0 $). Тогда $ a, b, c $ параллельны плоскости $ \pi(a, b) $, причём $ [a, b] \bot \pi(a, b) \Rightarrow ([a, b], c) = 0 $

### Достаточность

Пусть $ (a, b, c) = 0 $. Тогда либо $ |[a, b]| = 0 $, либо $ |c| = 0 $, либо $ cos \space \varphi = 0 $, где $ \varphi $ - угол между $ c $ и $ [a, b] $. Это значит, что либо $ a $ и $ b $ коллинеарны, либо $ c = 0 $, либо $ c || \pi(a, b) $. Во всех случаях $ a, b, c $ компланарны

### Теорема 2

Смешанное произведение некомпланарных векторов $ a, b, c $ равно по абсолютной величине объёму параллепипеда, построенного на данных векторах.

### Доказательство

Отложим все эти векторы из одной точки. Получится параллелепипед, рёбрами которого являются вектора. Обозначим за $ h $ высоту из конца вектора $ c $. $ S_{ab} = |[a, b]| $. Тогда $ V = S_{ab}h = ||[a, b]| \cdot |c| cos \space \varphi| = |(a, b, c)| $

### Теорема 3

Для любых векторов выполняется равенство $ \space ([a, b], c) = (a, [b, c]) $

### Доказательство

Пусть $ a, b, c $ не компланарны. Тогда тройки $ a, b, c $ и $ b, c, a $ одинаково ориентированы. Так как $ (a, [b, c]) = (b, c, a) $, из теоремы 2 вытекает изначатльное равенство.

### Следствия

1. Для любых векторов имеют место равенства $ (a, b, c) = (b, c, a) = (c, a, b) = -(b, a, c) = -(a, c, b) = -(c, b, a) $
2. Смешанное произведение линейно по каждому из сомножителей
