# 21. Метод Гаусса исследования и решения систем линейных алгебраических уравнений

Рассмотрим систему с верхней трапецевидной матрицей $ A $:  

$$
\begin{bmatrix}
    \begin{array}{ccccccc|c}
        a_{11} & a_{12} & \dots & a_{1r} & a_{1, r + 1} & \dots & a_{1n} & b_1 \\
        0      & a_{22} & \dots & a_{1r} & a_{2, r + 1} & \dots & a_{2n} & b_2 \\
        \dots & \dots   & \dots & \dots  & \dots        & \dots & \dots  & \dots \\
        0     & 0       & \dots & a_{rr} & a_{r, r + 1} & \dots & a_{rn} & b_r \\
        0     & 0       & \dots & 0      & 0            & \dots & 0      & b_{r + 1} \\
        \dots & \dots   & \dots & \dots  & \dots        & \dots & \dots  & \dots \\
        0     & 0       & \dots & 0      & 0            & \dots & 0      & b_m \\
    \end{array}
\end{bmatrix}
$$

Такая система совместна тогда и только тогда, когда $ b_k = 0, \ \forall k > r $

Пусть теперь система совместна. Реализуем все пункты общей теории исследования и решения совместной системы:

1. В качестве бащисного минора всегда можно взять минор, расположенный в левом верхнем углу.
2. Укороченная система состоит из первых $ r $ уравнений
3. Если $ r = n $, то система станет системой с треугольной матрицей  
$$
\begin{cases}
    a_{11}x_1 &+& a_{12}x_2 &+& \dots &+& a_{1r}x_r &=& b_1 \\
              &&  a_{22}x_2 &+& \dots &+& a_{1r}x_r &=& b_2 \\
              &&            &&  \dots &&  \dots     && \dots \\
              &&            &&        &&  a_{rr}x_r &=& b_r
\end{cases}
$$  
которая имеет единственное решение, которое легко найти, решая систему снизу вверх.
4. Если $ r < n $, то неизвестные $ x_{r + 1}, \dots, x_n $ будут свободными и система примет вид  
$$
\begin{cases}
    a_{11}x_1 &+& a_{12}x_2 &+& \dots &+& a_{1r}x_r &=& b_1 &-& a_{1, r + 1}x_{r + 1} &-& \dots &-& a_{1n} x_n  \\
              &&            &&  \dots &&  \dots     && \dots \\
              &&            &&        &&  a_{rr}x_r &=& b_r &-& a_{r, r + 1}x_{r + 1} &-& \dots &-& a_{rn} x_n
\end{cases}
$$  
Общее и частное решение системы находятся из получившейся системы с треугольной матрицей.

При помощи основного процесса мы можем привести матрицу $ A $ элементарными преобразованиями строк и перестановками столбцов к верхней треугольной форме. Если используемые при это преобразования применить ко всей расширенной матрице, то мы придём к системе, решения которой отличаются лишь нумерацией неизвестных.Метод Гаусса состоит в приведении системы уравнений к верхнему трапецевидному виду и исследовании получившейся системы. При этом, если при преобразовании системы использовались перестановки столбцов основной матрицы $ A $, то в полученных решениях необходимо восстановить исходную нумерацию неизвестных.
