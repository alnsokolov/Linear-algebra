# 17. Матрицы линейного оператора в различных базисах

Пусть $ e, t = eC $ - два базиса пространства $ V $ с матрицей перехода $ C $, а $ f, s = fD $ - два базиса пространства $ W $ с матрицей перехода $ D $. Одному и тому же линейному оператору соотвествуют матрицы $ A_{fe}, A_{st} $ в парах базисов $ e, f $ и $ t, s $ соотвественно.

### Теорема 77.3

Матрицы $ A_{fe} $ и $ A_{st} $ линейного оператора $ \mathcal{A \in L}(V, W) $ в различных парах базисов связаны соотношением

$$ A_{st} = D^{-1} A_{fe} C $$

### Доказательство

Для произвольного вектора $ x \in V $ имеем

$$ y_f = A_{fe} x_e, y_s = A_{st} x_t $$

В совю очередь $ x_e = C x_t, \space y_f = D y_s \Rightarrow y_f = A_{fe} x_e, y_s = A_{st} x_t \Leftrightarrow D y_s = A_{fe} C x_t $ или $ D A_{st} x_t = A_{fe} C x_t $. Так как это справедливо для любого $ x_t \in V $, $ D A_{st} = A_{fe} C $ или $A_{st} = D^{-1} A_{fe} C  $

Profit.

Следствия:
1. Матрицы оператора в различных базисах эквивалентны.
2. Ранг матрицы не зависит от выбора базиса.

### Теорема 77.4

Две матрицы над полем $ P $ размера $ m \times n $ эквивалентны тогда и только тогда, когда они являются матрицами одного и того же оператора $ \mathcal{A \in L}(V, W) $, где $ V, W $ - линейные пространства над полем $ P $ размерности $ n $ и $ m $ соотвественно.

### Доказательство

Возьмём такую пару матриц, так, что $ B = D^{-1} A C $. Также возьмём пару базисов пространств $ V $ и $ W $ $ e $ и $ f $. Поставим в соотвествие матрице $ A $ оператор $ \mathcal{A \in L}(V, W) $, имеющиё в паре базисов $ e, f $ матрицу $ A $. Тогда матрица $ B $ будет матрицей этого оператора в паре базисов $ t = eC, s = fD $.
Достаточность была рассмотрена в предыдущей теореме.
