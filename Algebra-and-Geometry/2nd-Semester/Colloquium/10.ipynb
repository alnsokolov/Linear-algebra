# 10. Ортонормированный базис и унитарные (ортогональные) матрицы

Базис, векторы которого образуют ортонормированную систему, называется ортонормированным базисом. $ e_1, \dots, e_k $ - ортонормированный базис $ E(U) $, если

$$ (e_i, e_j) = \begin{cases} 1, &i = j \newline 0, & i \neq j \end{cases} $$

В $n$-мерном евклидовом (унитарном) пространстве любая ортонормированная система из $n$ векторов образует базис.

Матрица $ U \in \mathbb{C}^{n \times n} $ называется унитарной, если $$ UU^H = U^HU = I, $$ матрица $ Q \in \mathbb{R}^{n \times n} $ называется ортогональной, если $$ QQ^T = Q^TQ = I $$.

### Теорема 70.5

Матрица перехода от ортонормированного базиса $ e $ к базису $ e' $ евклидова (унитарного) пространства ортогональна (унитарна) тогда и только тогда, когда $ e' $ - ортогональный базис.

### Доказательство

Столбцы матрицы перехода являются координатами векторов нового базиса в старом базисе. Их произведение $ (e_i', e_j') = \{ C^HC \}_{ij} = \sum_{k = 1}^n c_{ki} c_{kj} $ - другая формулировака ортогональности (унитарности) матрицы перехода. Следовательно матрица униатрна только в случае ортонормированностти базиса $ e' $.
