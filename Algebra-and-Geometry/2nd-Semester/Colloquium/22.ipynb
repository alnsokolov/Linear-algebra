# 22. Инвариантные подпространства. Индуцированный оператор.

Пусть $ \mathcal{A \in L}(V, V) $. Тогда подпространство $ L $ пространства $ V $ называется инвариантным относительно оператора $ \mathcal A $, если $ \forall x \in L \space\space \mathcal A x \in L $.

### Примеры

1. Тривиальные подпространства $ \{ \theta \}, V $
2. $ \ker \mathcal A, \text{im} \space \mathcal A $
3. Для $ \mathcal D $ это $ M_k, \space k = \overline{1, n - 1} $ для пространства $ M_n $

### Теорема 84.1

Пусть $ \mathcal{A \in L}(V, V) $ и $ L $ - нетривиальное инвариантное подпространство относительно $ \mathcal A $. Тогда существует базис подпространства $ \mathcal A $, в котором матрица имеет квазитреугольную форму.

### Доказаельство

Построим базис подпротсранства $ L \space e_1, \dots, e_k $ и расширим его до базиса всего пространства $ e_1, \dots, e_k, e_{k + 1}, \dots, e_n $. Строим как вырожаются вектора $ \mathcal A e_1, \dots, \mathcal A e_n $, строим матрицу, убеждаемся в её квазитреугольности.

Profit.

### Теорема 84.2

Если пространство $ V $ является прямой суммой нескольких инвариантных подпространств относительно оператора $ \mathcal A $, то есть базис пространства $ V $, в котором матрица оператора имеет квазидиагональную форму.

### Доказательство

Аналогично предыдущему, строим базис $ V $ из базисов инварианстных подпространств, рассматриваем образы векторов базиса, строим матрицу и убеждаемся в её квазидиагональности.

Profit.

Рассматривая линейный оператор только на инвариантном подпространстве, можно получить новый оператор (индуцированный оператор). Пусть $ L $ - протсранство, инвариантное относительно оператора $ \mathcal{A \in L}(V, V) $. Отображение $ \mathcal A|L: L \to L $, определённое равенством

$$ (\mathcal A | L)x = \mathcal x, \space \forall x \in L $$

называется индуцированным оператором, порождённым оператором $ \mathcal A $. Он совпадает с оператором на подпространстве $ L $ и не определён вне него. Итак, $ \mathcal A|L \in \mathcal L(L, L) $.
