# 25. Характеристический многочлен линейного оператора

Характеристичеким многочленом матрицы $ A \in P^{n \times n} $ называется функция

$$ f(\lambda) = \det (A - \lambda I), \space \lambda \in P $$

### Теорема 86.1

Характеристический многочлен матрицы $ A \in P^{n \times n} $ является многочленом $ n $-ой степени над полем $ P $.

### Доказательство

Каждый элемент матрицы $ A $ является многочленом от $ \lambda $ степени не выше 1, значит каждый член $ \det (A - \lambda I) $ является многочленом от $ \lambda $ степени не выше $ n $. Осталось показать, что эта степень в точности равна $ n $. Действиетльно, все все члены, отличные от $ (a_{11} - \lambda)(a_{22} - \lambda) \dots (a_{nn} - \lambda) $ имеют степень не выше $ n - 2 $. Значит коэффиценты перед $ \lambda^n $ и $ \lambda^{n - 1} $ определяются только с помощью члена $ (a_{11} - \lambda)(a_{22} - \lambda) \dots (a_{nn} - \lambda) $, который согласно формулам Виета имеет вид $ (-\lambda)^n + (a_{11} + \dots + a_{nn})(-\lambda)^{n - 1} + g_{n - 2}(\lambda) $, где $ g_{n - 2} $ - многочлен степени не выше $ n - 2 $ от переменной $ \lambda $. Значит наш искомый мночлен имеет степень, равную $ n - 2 $

Profit.

$ f(\lambda) = a_0 + a_1 (-\lambda) + \dots + a_n (-\lambda)^n $

$ a_k $ равен сумме всех миноров порядка $ k $ матрицы $ A $.

### Теорема 86.2

Характеристические многочлены подобных матриц совпадают.

### Доказательство

Допустим $ A = Q^{-1}BQ $. Тогда $ |A - \lambda I| = | Q^{-1}BQ -  Q^{-1}\lambda IQ| =  | Q^{-1}(B - \lambda I)Q| = |B - \lambda I| $

Profit.

### Теорема 86.3

Характеристический многочлен индуцированного оператора является делителем характеристического многочлена самого оператора.

### Доказательство

Следует из формы матрицы оператора с инвариантным подпространством.

### Теорема 86.4

Если $ V = L_1 \oplus L_2 \oplus \dots \oplus L_k $ - прямая сумма подпространств, инвариантных относительно оператора $ \mathcal{A \in L}(V, V) $, то характеристичекий многочлен этого оператора $ f(\lambda) $ равен произведению характиристических многочленов $ f_1(\lambda), f_2(\lambda), \dots, f_k(\lambda) $ индуцированных операторов $ \mathcal A|L_1, \mathcal A|L_2, \dots, \mathcal A|L_k $.

### Доказательство

Это следует из формы матрицы оператора $ \mathcal A $.

### Теоерма 86.5

Пусть $ V $ - линейное пространство над полем $ P $. Число $ \lambda \in P $ является собственным значением оператора $ \mathcal {A \in L}(V, V) $ тогда и только тогда, когда $ \lambda $ - корень его характеристического многочлена.

### Доказательство

По определению $ \lambda $ является собственным значением оператора тогда и только тогда, когда $ \exists x \in V: $

$$ \begin{cases} \mathcal Ax = \lambda x \newline x \neq \theta \newline \lambda  \in P \end{cases} $$

Это равносильно вырожденности оператора $ \mathcal {A - \lambda I} $ при некотором $ \lambda \in P $ или 

$$ \det (\mathcal{A - \lambda I}) = 0 $$

при некотором $ \lambda \in P $. Таким образом число $ \lambda $ является собственным значением оператора $ \mathcal A $ тогда и толкьо тогда, когда оно является корнем его характеристического многочлена.

Profit.

Уравнение $ \det (\mathcal{A - \lambda I}) = 0 $ характеристическим уравнением оператора $ \lambda A $, а корни характеристического многочлена называются собственными числами оператора $ \mathcal A $.
