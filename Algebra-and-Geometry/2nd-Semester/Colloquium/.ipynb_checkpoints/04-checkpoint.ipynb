# 4. Прямая сумма подпространств

Сумма подпространств линейного пространства называется прямое, если разложение каждого вектора в ней по слагаемым подпространствам единственно. Обозначение: $ L_1 \oplus \dots \oplus L_k $

### Тоерема 66.1 (критерий прямой суммы)

Для подпространств $ L_1, \dots, L_k $ конечномерного пространства $ V $ следующие утверждения равносильны
1. Сумма подпространств - прямая
2. Совокупность базисов подпространств линейно независима
3. Совокупность базисов подпространств образует базис суммы
4. $ \dim \sum_{i = 1}^k L_i = \sum_{i = 1}^k \dim L_i $
5. Существует вектор $ a $, принадлежащий сумме, для которого разложение по подпространствам единственно
6. Произвольная сиситема ненулевых векторов $ a_1, \dots, a_k $, взятых по одному из каждого подпространства, линейно независима
7. $ L_1 \cap L_2 = \{ \theta \} $

### Доказательство

$ 1 \Rightarrow 2 $ Допустим совокупность базисов линейного подпространства линейно зависима. Тогда существует как минимум два разложения вектора $ \theta $ по подпространствам.  

$ 1 \Leftarrow 2 $ Допустим сумма не прямая. Тогда существует вектор, который можно разложить как минимум двумя способами по подпространствам. Рассмотрим их и получим разложение $ \theta $ двумя способами, значит совокупность базисов линейно зависима.  

$ 2 \Leftrightarrow 3 $ Следует из теоремы 65.2 (билет 3).

$ 3 \Leftrightarrow 4 $ Утверждения отличабтся только терминологией.

$ 1 \Rightarrow 5 $ Очевидно

$ 1 \Leftarrow 5 $ Получим два разложения нуля, сложим с вектором $ a $ и получим два разложения $ a $

$ 1 \Rightarrow 6 $ Пусть система $ a_1, \dots, a_k $ линейно зависима. Тогда существует ещё одно разложение нуля по подпространствам и сумма получается не прямой

$ 1 \Leftarrow 6 $ Пусть сумма не прямая. Тогда есть вектор, который можно разложить по подпростространствам двумя различными способами, вычисть уравнения и получить разложение $ \theta $ через ненулевые вектора подпространств, что противоречит 6

$ 4 \Leftrightarrow 7 $ Следует из теоремы 65.3 (билет 3)

### Теорема 66.2

Линейное пространство $ V $ является прямой суммой двух своих подпространств $ L_1 $ и $ L_2 $ тогда и только тогда, когда:
1. $ \dim V = \dim L_1 + \dim L_2 $
2. $ L_1 \cap L_2 = \{ \theta \} $

### Доказательство

Необходимость следует из теоремы 66.1

Достаточность. Пусть $ L = L_1 \oplus L_2 \Rightarrow \dim L = \dim L_1 + \dim L_2 = \dim V $. Значит $ L = V = L_1 \oplus L_2 $


ч.т.д.

Дополнительным подпространством относительно подпространства $ L $ линейного протсранства $ V $ называется подпространство $ L^\delta $, такое, что $ L \oplus L^\delta = V $ 

### Теорема 66.3

Для любого подпространства $ L $ линейного пространства $ V $ существует дополнительное подпространство.

### Доказательство

Получим базис $ L $, дополним его до $ V $, линейная оболочка векторов, которыми мы дополняли и дополнительное пространство.
