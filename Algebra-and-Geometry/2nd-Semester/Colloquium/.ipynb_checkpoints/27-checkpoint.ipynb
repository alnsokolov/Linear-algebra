# 27. Собственное подпространство. Алгебраическая и геометрическая кратности собственного подпростанства.

$ W_{\lambda_i} = \{ x \in V | A x = \lambda_i x \} $ - собственное подпространство оператора $ \mathcal A $, соотвествуещее собственному значению $ \lambda_i $. $ W_{\lambda_i} = \ker (A - \lambda_i I) \Rightarrow $ соственное подпространство - подпространство $ V $. $ W_{\lambda_i} $ состоит из вектора $ \theta $ и всех собственных векторов оператора $ \mathcal A $, соотвествующих собственному значению $ \lambda_i $. $ W_{\lambda_i} $ инвариантно относительно оператора $ \mathcal A $. $ \dim W_{\lambda_i} $ - геометрическая кратность собственного значения $ \lambda_i $.

### Тоерема 87.1

Геометрическая кратность не превышает алгебраическую.

### Доказательство

Пусть $ s $ и $ m $ - геометрическая и алгебраическая кратности соотвественно для собственного значения $ \lambda_0 $. Индуцированный оператор $ \mathcal A | W_{\lambda_0} $ имеет базис $ e_1, \dots, e_s $. Тогда матрицей данного оператора в данном базисе будет диагональная маттрица с $ \lambda_0 $ на диагонали. Значит характеристический многочлен этого индуцированного оператора будет равен $ (\lambda - \lambda_0)^s $. Он является делителем характеристического многочлена самого оператора, но, т.к. $ (\lambda - \lambda_0)^m $ является множителем этого многочлена, $ s \leq m $

Profit.

### Теорема 87.2

Сумма собственных подпротсранств оператора, отвечающих различным собственным значениям, является прямой суммой.

### Доказательство

Пусть $ \lambda_1, \dots, \lambda_p $ - попарно различные собственные значения оператора. Тогда для $ W_{\lambda_1}, \dots, W_{\lambda_p} $ выполняется условие прямой суммы: любая система из нулевых векторов, взятых по одному из каждого подспространства, линейно независима, как система собственных векторов, соотсветствующих разным собственным значениям.
