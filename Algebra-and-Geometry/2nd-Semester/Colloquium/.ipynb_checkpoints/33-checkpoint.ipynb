# 33. Корневые векторы. Канонический базис корневого подпространства.

Пусть $ \lambda_i $ - собственное значение оператора $ \mathcal A $. Вектор $ x \in V $ называется корневым вектором оператора $ \mathcal A $, отвечающим собственному значению $ \lambda_i $, если $ (\mathcal A - \lambda_i I)^k x = \theta $ при некотором $ k \in \mathbb{Z}, k \geq 0 $. Высотой корнового дерева называют минимальное значение $ k $, удовлетворяющее заданным условиям. Очевидно, что высота нулевого вектора (и только его) равна 0.

#### Простейшие свойства

1. Корневые векторы высоты $1$ являются собственными векторами оператора
2. Если $ x $ - корневой вектор высоты $ k > 0 $, то вектор $ (\mathcal A - \lambda_j I)x $ является корневым вектором высоты $ k - 1 $
3. Корневые векторы различных высот линейно независимы

Корневые векторы высоты $ k > 1 $ называются присоединёнными векторами $ k - 1 $ порядка.

Множество всех корневых векторов оператора $ \mathcal A $, отвечающих собственному значению $ \lambda_j $ называется корневым подпространством оператора $ \mathcal A $, отвечающих собственному значению $ \lambda_j $. Обозначение: $ K_{\lambda_j} $. Итак:

$$ K_{\lambda_j} = \{ x \in V | \exists k \in \mathbb{Z}, \space k \geq 0: (\mathcal A - \lambda_j I)^k x = \theta \} $$

### Канонический базис корневого подпространства

Пусть $ K_{\lambda_j} $ - корневое подпространство оператора $ \mathcal A $, отвечающее собственному значению $ \lambda_j $. Положим 

$$ \mathcal{B = A - \lambda_j I}, N_k = \ker \mathcal B^k, n_k = \dim N_k, r_k = \text{rg}\space \mathcal B^k $$

Построим само корневое пространство. Для этого надо найти минимальное $ q \in \mathbb{N}: N_{q} = N_{q + 1} = K_{\lambda_j} $. При этом имеем $ n_1 = s_j < n_2 < \dots < n_q = m_j $, где $ s_j $ и $ m_j $ - геометрическая и алгебраическая кратности $ \lambda_j $.

Теперь будем строить базис $ K_{\lambda_j} $, последовательно просматривая $ N_q, N_{q - 1}, \dots, N_1 $.

$ N_q $) Векторы $ f_1, \dot, f_{t_q} $ - векторы, дополняющие базис $ N_{q - 1} $ до $ N_q $. Ясно, что:  
- они будут корневыми высоты $ q $  
- их количество равно $ n_q - n_{q - 1} $  
- $ t_q = n_q - n_{q - 1} = (n_q - n_{q - 1}) - (n_{q + 1} - n_q) = -n_{q + 1} + 2n_q - n_{q - 1} $, т.к. $ n_{q + 1} = n_q $
- Никакая нетривиальная линейная комбинация этих векторов не принадлежит $ N_{q - 1} $ (назовём их линейно независимыми над $ N_{q - 1} $)
$ N_{q - 1} $) Построим векторы $ \mathcal B f_1, \mathcal B f_2, \dots, \mathcal B f_{t_q} $ эти векторы являются корневыми высоты $ q - 1 $ и они линейно независимы над $ N_{q - 2} $. Дополним их векторами $ g_1, \dots, g_{t_{q - 1}} \in N_{q - 1} $ так, чтобы вектора $  \mathcal B f_1, \mathcal B f_2, \dots, \mathcal B f_{t_q}, g_1, \dots, g_{t_{q - 1}} $ дополняли произвольный базис $ N_{q} $ до базиса $ N_{q - 1} $. Выполняем такие же построения до пространства $ N_1 $.

В итоге получаем вектора $  \mathcal B^{q - 1} f_1, \mathcal B^{q - 1} f_2, \dots, \mathcal B^{q - 1} f_{t_q},  \mathcal B^{q - 2} g_1, \mathcal B^{q - 2} g_2, \dots, \mathcal B^{q - 2} g_{t_{q - 1}}, \dots, u_1, \dots, u_{t_1} $. Эти вектора:
- являются собственными векторами
- их количество равно $ n_1 $
- они линейно независимы

Полученные вектора для удобства объдиняются в таблицу котоая называется Жордоновой лестницей.

### Теорема 92.1

Построенная система векторов образует базис корневого подпротсранства $ K_{\lambda_j} $.

### Доказательство

Количество векторов в полученной системе равно $ n_1 + (n_2 - n_1) + \dots  + (n_q - n_{q - 1}) = n_q = \dim K_{\lambda_j} $ и они все линейно независимые.
