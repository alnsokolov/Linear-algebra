# 28. Операторы простой структуры. Критерии оператора простой структуры.

Критерий оператора простой структуры. Линейный оператор $ \mathcal{A \in L}(V, V) $ нызвается оператором простой структуры, если в пространстве $ V $ существует бызис из его собственных векторов.

### Теорема 88.1

Линейный оператор $ \mathcal{A \in L}(V, V) $ является оператором простой структуры тогда и только тогда, когда в протсранстве $ V $ существует бызис $e $, в котором его матрица диагональна.
 
### Доказательство

Пусть $ \dim V = n $. Оператор $ \mathcal A $ является оператором простой структуры тогда и только тогда, когда он имеет $ n $ линейно независимых собственных векторов $ e_1, \dots, e_n $. Это равносильно существованию базисы, в котором $ A_e = \text{diag} (\lambda_1, \dots, \lambda_n) $, что повторяет формулировку теоремы.

Profit.

Следствие. В пространстве $V$ размерности $n$, если $ A $ имеет $ n $ различных собственных значений, то $ A $ - оператор простой структуры.

### Теорема 88.2

Линейный оператор $ \mathcal{A \in L}(V, V) $ имеет простую структуру тогда и только тогда, когда все его собстыенные подпространства в прямой сумме дают всё пространство $ V $.

### Доказательство

Необходимость. Говорим, что сумма подпространств является вложена в пространство $ V $ (очевидно). Потом берём базис из собственных векторов $ e_1, \dots, e_n $, говорим, что любой вектор, лежащий в $ V $ раскладывается по этим подпротсранствам, значит $ V $ вложено в сумму подпространств. Следовательно $ W_{\lambda_1} + W_{\lambda_p} = V $ и эта сумма прямая.

Достаточность. Совокупность базисов слагаемых суммы пространств даёт базис самой суммы, значит в протсранстве $ V $ есть базис из собственных векторов/базисов собственных подпротсранств.

Profit.

### Теорема 88.3

Линейный оператор, действующий в комплексном пространстве, имеет простую структуру тогда и только тогда, когда для каждого его собственного значения геометрическая кратность совпадает с алгебраической.

### Доказательство

Пусть $ \lambda_1, \dots, \lambda_p $ - попарно различные собственные значения. А $ m_k, s_k $ - алгебраическа и собственная кратность, соотсвествующие $ \lambda_k $. Тогда $ \dim V = m_1 + \dots + m_p $, но $ s_k \leq m_k, \space k = \overline{1, p} \Rightarrow s_k = m_k, \space k = \overline{1, p} $

#### Матричная формулировака

Всё то же самое но заменяем слово "оператор" на слово "матрица". Важный факт - собственные значения оператора $ \mathcal A $ и его матрицы $ A_e $ в произвольном базисе $ e $ совпадают, а собственные векторы матрицы $ A_e $ являются координатными столбцами собственных векторов оператора $ \mathcal A $.

### Теорема 88.4

Матрица имеет простую структуру тогда и только тогда, когда она подобна диагональной матрице.

### Доказательство

Строим наш матрицу, фиксируем базис, гвоорим, что матрица совпадает с матрицей оператора в этом базисе. Потом говорим, что матрица имеет простую структуру, тогда и только тогда, когда оператор имеет простую структуру, значит тогд аи только тогда, когда существует базис, в котором матрица оператора имеет диагональную форму. Ну а последнее равносильно подобие матриц.

Profit.

Важный пример матрицы, которая не является матрицей простой структуры, приведён в учебнике и имя ему - Жорданова клетка.
