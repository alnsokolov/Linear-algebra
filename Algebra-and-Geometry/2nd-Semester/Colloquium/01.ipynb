# 1. Линейное пространство над произвольным полем. Ранг и база системы векторов.

Дано поле $ P $. Непустое множество $ V $ называется линейным пространством над полем $ P $, если на этом поле определены внутрениий и неший законым композиции $ V \times V \to V $ и $ P \times V \to V $, называемые сложением и умножением соотвественно, удовлетворяющие следующим аксиомам для любых $ a, b, c \in V, \space \alpha, \beta \in P $
1. $ a + b = b + a $
2. $ (a + b) + c = a + (b + c) $
3. $ \exists \theta \in V: a + \theta = \theta + a = a $
4. $ \forall a \in V \space \exists (-a) \in V: a + (-a) = (-a) + a = \theta $
5. $ 1 \cdot a = a $
6. $ \alpha (\beta a) = (\alpha \beta) a $
7. $ (\alpha + \beta)a = \alpha a + \beta a $
8. $ \alpha (a + b) = \alpha a + \alpha b $

Линейное пространство над полем $ \mathbb{R} $ называется вещественным, над полем $ \mathbb{C} $ - комплексным. Вектора $ a, b \in V $ называются коллинеарными, если $ a = \alpha b, \alpha \in P $

Аффинным (или точечно-аффинным) пространством над линеным пространством $ V $ называется множество $ S $ элементов, называемых точками, для которого заданы
1. Линейное пространство $ V $ над полем $ P $
2. Отображение $ v: S \times S \to V $, ставящее в соотвествие кажой паре точек $ A, B \in S $ вектор $ v(A, B) \in V $ и удовлетворяющее следующим аксиомам:
    - $ \forall A \in S, a \in V \space \exists! B \in S: \space\space v(A, B) = a $
    - $ \forall A, B, C \in S \space\space v(A, B) + v(B, C) = v(A, C) $.

Вектор $ v(A, B) $ обозначается как $ \vec{AB} $

Из аксиом аффинного пространства соледует, что
1. $ \vec {AB} = \theta \Leftrightarrow A = B $
2. $ \vec{AB} = -\vec{BA} $

### Примеры

1. Геометрические пространства $ V_n $ являются аффинными пространствами над самими собой \.
2. Любое линейное пространство можно рассматривать как афинное. Для этого достаточно веторы назвать точками и определить отображение $ v: V \times V \to V $ как $ v(a, b) = a - b $
3. Люое аффинное пространство можно рассматривать как линейное. Для этого зафиксируем точку пространства и назовыём её началом. Каждой точке пространства поставим в соотвествие её радиус-вектор относительно этой точки. Их множество и составляет линейное пространство $ V $.

### Теорема 60.1

Отображение $ \varphi: S \to V: \space\space \varphi(A) = \vec{OA} $ является биекцией, сохраняющей операцию $ v $ аффинного пространства, т.е $ v(A, B) = v(\varphi(A), \varphi(B)) $

### Доказательство

Биектикность следует непосредственно из аксиомы 1. $ v(A, B) = v(\varphi(A), \varphi(B)) $. Согласно определению аффинного пространства через линейное (в примере 2), $ v(A, B) = v(\varphi(A), \varphi(B)) \Leftrightarrow \vec{AB} = \vec{OB} - \vec{OA} $.

Данная теорема помогает отождествить линейные и афинные пространства. Единственное различие - в линейном существует единственный вектор $ \theta $, а в аффинном в качестве начала координат можно взять любую точку.

## Ранг и база системы векторов

Будем рассматривать конечные системы векторов $ a_1, \dots, a_k $ линейного пространства $ V $. Линейно независимая подсистема системы векторов, через которую выражается любой вектор системы, называется базой этой системы векторов.

### Теорема 61.1

Подсистема системы векторов является базой этой системы векторов тогда и только тогда, когда образует максимальную линейно независимую систему.

### Доказательство

Небходимость. Берём базу, рассматриваем, что будет, если мы возьмём большую подсистему.  
Достаточность. Берём максимальную линейно независимую систему векторов, любая большая система векторов будет линейно зависимой $ \Rightarrow $ данная система будет базой.

Следствие - все базы состоят из одинакового числа векторов.

Число векторов в базе называется рангом системы и обозначается как $ rg(a_1, \dots, a_k) $

### Теорема 61.2

Если система векторов $ a_1, \dots, a_k $ линейно выражается через $ b_1, \dots, b_m $, то $ rg(a_1, \dots, a_k) \leqslant rg(b_1, \dots, b_m) $

### Доказательство

Пусть $ a_1, \dots, a_r $ и $ b_1, \dots, b_s $ - базы рассматривыемых систем. Тогда $ s \geqslant r $, иначе, т.к. база первой системы выражется через базу второй, в силу теоремы 16.2 (матрицы), $ a_1, \dots, a_r $ была бы зависима.

Следствие 2. Ранги эквивалентных систем совпадают.
Следствие 3. Эквивалентные линейно независимые системы векторов состоят из одного и того же числа векторов.
