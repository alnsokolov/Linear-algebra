# 26. Условие существования собственных векторов линейного оперратора. Собственные веторы линейного оператора в комплексном пространстве.

### Теоерма 86.5

Пусть $ V $ - линейное пространство над полем $ P $. Число $ \lambda \in P $ является собственным значением оператора $ \mathcal {A \in L}(V, V) $ тогда и только тогда, когда $ \lambda $ - корень его характеристического многочлена.

### Доказательство

По определению $ \lambda $ является собственным значением оператора тогда и только тогда, когда $ \exists x \in V: $

$$ \begin{cases} \mathcal Ax = \lambda x \newline x \neq \theta \newline \lambda  \in P \end{cases} $$

Это равносильно вырожденности оператора $ \mathcal {A - \lambda I} $ при некотором $ \lambda \in P $ или 

$$ \det (\mathcal{A - \lambda I}) = 0 $$

при некотором $ \lambda \in P $. Таким образом число $ \lambda $ является собственным значением оператора $ \mathcal A $ тогда и толкьо тогда, когда оно является корнем его характеристического многочлена.

Profit.

1. $ \det (\mathcal{A - \lambda I}) = 0 $ - характеристическое уравнение оператора $ \mathcal A $, $ \lambda_i $ - характеристичсекие числа оператора
2. Существование собственных векторов сводится к сществованию корней храктеристического многочлена. В алгебраическом замкнутом поле $ \mathbb C $ комплексных числе у любого многочлена степени $ n \geq 1 $ имеется $ n $ корней. Значит:

### Теорема 86.6

Произвольный линейный оператор $ \mathcal A $ в комплексном линейном пространстве имеет:
- $n$ собстченных значений (считая кратность их как корней многочлена)
- хотя бы 1 собстыенный вектор
- в любом его инвариантном пространстве имеется хотя бы один собственный вектор

Метод нахождения собственных векторов оператора $ \mathcal A $:
1. Строим характеристический многочлен оператора
2. Ищем его корни, т.е. $ \lambda_i \in P $ - собственное значение $ \mathcal A $
3. Векторы $ \ker (\mathcal{A - \lambda_i I}) $ - собственыне векторы для $ \lambda_i $  
   Если $ e_1, \dots, e_n $ - базис $ V $, то ФСР $ (A - \lambda_i I)x = 0 $ и есть набор линейно независимых собственных векторов для $ \lambda_i $ в координатах $ e $.
